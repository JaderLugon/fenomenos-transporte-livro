## 🏭 Descrição do Problema

Uma indústria metalúrgica necessita de um sistema hidráulico para elevar uma carga de 50 toneladas a uma altura de 3 m em 10 segundos.

### Especificações

- **Carga:** $W = 50$ toneladas $= 490.500$ N
- **Curso do cilindro:** $L = 3$ m
- **Tempo de elevação:** $t = 10$ s
- **Pressão máxima do sistema:** $p_{max} = 210$ bar
- **Fluido:** Óleo hidráulico ISO VG 46

### Propriedades do Óleo (40°C)

- $\rho = 870$ kg/m³
- $\mu = 0,046$ Pa·s
- $\nu = 5,3 \times 10^{-5}$ m²/s

```
# ============================================================
# DADOS DO SISTEMA
# ============================================================

# Especificações mecânicas
W = 50e3 * 9.81    # N (peso da carga)
L_curso = 3.0      # m (curso do cilindro)
t_elev = 10.0      # s (tempo de elevação)
p_max = 210e5      # Pa (pressão máxima = 210 bar)

# Propriedades do óleo
rho_oleo = 870     # kg/m³
mu_oleo = 0.046    # Pa·s
nu_oleo = 5.3e-5   # m²/s

print("📊 Especificações do sistema hidráulico:")
print(f"  • Carga: W = {W/1000:.1f} kN ({W/9.81/1000:.0f} toneladas)")
print(f"  • Curso: L = {L_curso} m")
print(f"  • Tempo de elevação: t = {t_elev} s")
print(f"  • Pressão máxima: p_max = {p_max/1e5:.0f} bar")
print(f"  • Fluido: Óleo ISO VG 46 (μ = {mu_oleo*1000:.1f} mPa·s)")
```

## 🧮 Dimensionamento do Cilindro Hidráulico

### Cálculo da Área do Pistão

```
# ============================================================
# DIMENSIONAMENTO DO CILINDRO
# ============================================================

print("=" * 60)
print("DIMENSIONAMENTO DO CILINDRO HIDRÁULICO")
print("=" * 60)

# Área do pistão (considerando p_max)
A_pistao = W / (p_max * 0.85)  # 85% da pressão máxima (margem de segurança)

# Diâmetro do pistão
D_pistao = np.sqrt(4 * A_pistao / np.pi)

# Diâmetro comercial (arredondar para cima)
D_pistao_com = np.ceil(D_pistao * 1000 / 25) * 25 / 1000  # múltiplos de 25 mm
A_pistao_com = np.pi * D_pistao_com**2 / 4

print(f"\n🧮 Cálculo da área do pistão:")
print(f"  • A = W / (p_max × 0,85) = {W/1000:.1f}×10³ / ({p_max/1e5:.0f}×10⁵ × 0,85)")
print(f"  • A = {A_pistao:.4f} m²")
print(f"  • D = √(4A/π) = {D_pistao*1000:.1f} mm")
print(f"\n📏 Diâmetro comercial adotado: D = {D_pistao_com*1000:.0f} mm")
print(f"  • Área real: A = {A_pistao_com:.4f} m²")

# Pressão de operação
p_op = W / A_pistao_com

print(f"\n🔍 Pressão de operação:")
print(f"  • p_op = W/A = {W/1000:.1f}×10³ / {A_pistao_com:.4f}")
print(f"  • p_op = {p_op/1e5:.1f} bar ({p_op/p_max*100:.1f}% da p_max)")
```

## 💧 Dimensionamento da Vazão

### Vazão Necessária

```
# ============================================================
# VAZÃO NECESSÁRIA
# ============================================================

print("\n" + "=" * 60)
print("DIMENSIONAMENTO DA VAZÃO")
print("=" * 60)

# Volume do cilindro
V_cilindro = A_pistao_com * L_curso

# Vazão necessária
Q_nec = V_cilindro / t_elev

print(f"\n🧮 Cálculo da vazão:")
print(f"  • Volume: V = A × L = {A_pistao_com:.4f} × {L_curso} = {V_cilindro:.4f} m³")
print(f"  • Vazão: Q = V/t = {V_cilindro:.4f} / {t_elev} = {Q_nec:.5f} m³/s")
print(f"  • Q = {Q_nec*60*1000:.2f} L/min")

# Seleção da bomba (com margem de 20%)
Q_bomba = Q_nec * 1.2

print(f"\n🔋 Vazão da bomba (com margem de 20%):")
print(f"  • Q_bomba = {Q_bomba*60*1000:.2f} L/min")
```

## 🔧 Dimensionamento da Tubulação

### Diâmetro da Tubulação (Velocidade Recomendada)

Para óleo hidráulico, a velocidade recomendada é:
- **Linha de pressão:** 3-6 m/s
- **Linha de sucção:** 0,5-1,5 m/s
- **Linha de retorno:** 1,5-3 m/s

```
# ============================================================
# DIMENSIONAMENTO DA TUBULAÇÃO
# ============================================================

print("=" * 60)
print("DIMENSIONAMENTO DA TUBULAÇÃO")
print("=" * 60)

# Velocidade recomendada para linha de pressão
V_rec = 4.0  # m/s (valor típico)

# Diâmetro interno
D_int = np.sqrt(4 * Q_bomba / (np.pi * V_rec))

# Diâmetro comercial (tubo de aço)
D_int_com = np.ceil(D_int * 1000 / 2) * 2 / 1000  # múltiplos de 2 mm
D_ext_com = D_int_com + 0.004  # parede de 2 mm

print(f"\n🧮 Cálculo do diâmetro:")
print(f"  • D_int = √(4Q/(πV)) = √(4×{Q_bomba*60*1000:.2f}×10⁻³/60 / (π×{V_rec}))")
print(f"  • D_int = {D_int*1000:.1f} mm")
print(f"\n📏 Diâmetro comercial adotado:")
print(f"  • D_int = {D_int_com*1000:.0f} mm")
print(f"  • D_ext = {D_ext_com*1000:.0f} mm (parede de 2 mm)")

# Velocidade real
A_tubo = np.pi * D_int_com**2 / 4
V_real = Q_bomba / A_tubo

print(f"\n🔍 Velocidade real:")
print(f"  • V = Q/A = {Q_bomba*60*1000:.2f}×10⁻³/60 / {A_tubo:.5f} = {V_real:.2f} m/s")
```

## 🌊 Verificação do Regime de Escoamento

### Número de Reynolds

```
# ============================================================
# REGIME DE ESCOAMENTO
# ============================================================

print("=" * 60)
print("VERIFICAÇÃO DO REGIME DE ESCOAMENTO")
print("=" * 60)

Re = V_real * D_int_com / nu_oleo

print(f"\n🔍 Número de Reynolds:")
print(f"  • Re = V·D/ν = {V_real:.2f} × {D_int_com*1000:.0f}×10⁻³ / {nu_oleo:.2e}")
print(f"  • Re = {Re:.0f}")

if Re < 2000:
    regime = "LAMINAR"
    print(f"\n✅ Regime: {regime} (Re < 2000)")
    print(f"   Aplicável: Hagen-Poiseuille")
elif Re > 2400:
    regime = "TURBULENTO"
    print(f"\n⚠️ Regime: {regime} (Re > 2400)")
    print(f"   Aplicável: Darcy-Weisbach")
else:
    regime = "TRANSIÇÃO"
    print(f"\n⚠️ Regime: {regime} (2000 < Re < 2400)")
```

## 📉 Perda de Carga (Hagen-Poiseuille)

### Cálculo da Perda de Carga em Regime Laminar

```
# ============================================================
# PERDA DE CARGA (HAGEN-POISEUILLE)
# ============================================================

print("=" * 60)
print("PERDA DE CARGA - HAGEN-POISEUILLE")
print("=" * 60)

# Comprimento equivalente da tubulação (incluindo acessórios)
L_tubo = 10.0  # m (estimativa)
K_acessorios = 10.0  # coeficiente total estimado
L_eq = L_tubo + K_acessorios * D_int_com

print(f"\n📊 Dados da tubulação:")
print(f"  • Comprimento real: L = {L_tubo} m")
print(f"  • Comprimento equivalente: L_eq = {L_eq:.1f} m (com acessórios)")

# Perda de carga (Hagen-Poiseuille)
hf = 32 * mu_oleo * L_eq * V_real / (rho_oleo * 9.81 * D_int_com**2)

print(f"\n🧮 Cálculo (Hagen-Poiseuille):")
print(f"  • h_f = 32·μ·L·V / (ρ·g·D²)")
print(f"  • h_f = 32 × {mu_oleo} × {L_eq:.1f} × {V_real:.2f} / ({rho_oleo} × 9,81 × {D_int_com}²)")
print(f"  • h_f = {hf:.3f} m")

# Perda de pressão
delta_p = rho_oleo * 9.81 * hf

print(f"\n📊 Perda de pressão:")
print(f"  • Δp = ρ·g·h_f = {delta_p/1e5:.2f} bar")
print(f"  • Representa {delta_p/p_op*100:.2f}% da pressão de operação")
```

## 🔋 Potência do Sistema

### Potência Hidráulica e Elétrica

```
# ============================================================
# POTÊNCIA DO SISTEMA
# ============================================================

print("=" * 60)
print("POTÊNCIA DO SISTEMA")
print("=" * 60)

# Potência hidráulica
P_hid = p_op * Q_bomba

print(f"\n🔋 Potência hidráulica:")
print(f"  • P_hid = p_op × Q = {p_op/1e5:.1f}×10⁵ × {Q_bomba:.5f}")
print(f"  • P_hid = {P_hid/1000:.2f} kW")

# Potência mecânica (elevação da carga)
P_mec = W * L_curso / t_elev

print(f"\n🔋 Potência mecânica (útil):")
print(f"  • P_mec = W × L / t = {W/1000:.1f}×10³ × {L_curso} / {t_elev}")
print(f"  • P_mec = {P_mec/1000:.2f} kW")

# Eficiência mecânica
eta_mec = P_mec / P_hid

print(f"\n📊 Eficiência mecânica:")
print(f"  • η_mec = P_mec / P_hid = {P_mec/1000:.2f} / {P_hid/1000:.2f} = {eta_mec*100:.1f}%")

# Potência elétrica (considerando η_bomba = 0,85 e η_motor = 0,90)
eta_bomba = 0.85
eta_motor = 0.90
eta_total = eta_bomba * eta_motor

P_el = P_hid / eta_total

print(f"\n🔋 Potência elétrica:")
print(f"  • η_total = η_bomba × η_motor = {eta_bomba} × {eta_motor} = {eta_total:.2f}")
print(f"  • P_el = P_hid / η_total = {P_hid/1000:.2f} / {eta_total:.2f} = {P_el/1000:.2f} kW")
print(f"  • Motor recomendado: {P_el/1000*1.2:.1f} kW (com margem de 20%)")
```

## 💰 Análise Econômica

### Custo Operacional

```
# ============================================================
# ANÁLISE ECONÔMICA
# ============================================================

print("=" * 60)
print("ANÁLISE ECONÔMICA")
print("=" * 60)

# Parâmetros operacionais
ciclos_dia = 50      # ciclos/dia
dias_ano = 250       # dias/ano
tempo_ciclo = 30     # s/ciclo (inclui retorno)
tarifa = 0.90        # R$/kWh

# Energia por ciclo
E_ciclo = P_el * tempo_ciclo / 3600  # kWh

# Energia anual
E_ano = E_ciclo * ciclos_dia * dias_ano

# Custo anual
custo_ano = E_ano * tarifa

print(f"\n📊 Parâmetros operacionais:")
print(f"  • Ciclos: {ciclos_dia}/dia × {dias_ano} dias/ano = {ciclos_dia*dias_ano} ciclos/ano")
print(f"  • Tempo por ciclo: {tempo_ciclo} s")
print(f"  • Tarifa: R$ {tarifa:.2f}/kWh")

print(f"\n💰 Custo energético:")
print(f"  • Energia por ciclo: E = {P_el/1000:.2f} × {tempo_ciclo}/3600 = {E_ciclo:.4f} kWh")
print(f"  • Energia anual: E = {E_ciclo:.4f} × {ciclos_dia*dias_ano} = {E_ano:.1f} kWh")
print(f"  • Custo anual: C = {E_ano:.1f} × {tarifa:.2f} = R$ {custo_ano:.2f}/ano")
```

### Observações Importantes

1. **Regime laminar:** Devido à alta viscosidade do óleo, o escoamento é laminar, o que resulta em perdas de carga moderadas mesmo em tubulações longas.

2. **Eficiência energética:** A eficiência mecânica do sistema é ~85%, típica de sistemas hidráulicos bem projetados.

3. **Manutenção:** Sistemas hidráulicos exigem manutenção regular (troca de óleo, filtros, vedação) para garantir confiabilidade.

4. **Segurança:** Pressões elevadas (176 bar) exigem componentes de alta qualidade e válvulas de segurança.

### Possíveis Melhorias

- **Acumulador hidráulico:** Armazena energia para picos de demanda, reduzindo potência da bomba
- **Bomba de vazão variável:** Ajusta vazão conforme demanda, economizando energia
- **Sistema de resfriamento:** Mantém temperatura do óleo em faixa ótima (40-60°C)
