## 📋 Resumo

Dimensionamos um dissipador térmico para um processador que dissipa 120 W. O dissipador possui 8 aletas retangulares de alumínio.

**Objetivos:**
1. Calcular a eficiência de cada aleta
2. Determinar o calor dissipado por aleta
3. Estimar a temperatura da base
4. Verificar se atende ao limite térmico do processador

---

## 🎯 Objetivos de Aprendizagem

- Aplicar eficiência de aletas retangulares
- Calcular calor dissipado por múltiplas aletas
- Analisar restrições térmicas de componentes eletrônicos

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
print('✅ Ambiente OK')
```

---

## 💻 Seção 1: Dados do Processador

| Parâmetro | Valor |
|-----------|-------|
| Potência dissipada | 120 W |
| Temperatura ambiente | 25°C |
| Temperatura máxima do chip | 85°C |
| Resistência térmica chip-base | 0,2 °C/W |
| Material das aletas | Alumínio (k = 205 W/(m·K)) |
| Número de aletas | 8 |
| Dimensões da aleta | L = 40 mm, t = 3 mm, w = 60 mm |
| Coef. convectivo (ar) | 50 W/(m²·K) |

```python
# Dados
Q_total = 120        # W
T_inf = 25           # °C
T_chip_max = 85      # °C
R_chip_base = 0.2    # °C/W
k_al = 205           # W/(m*K)
N_aletas = 8
L = 0.040
t = 0.003
w = 0.060
h = 50               # W/(m^2*K)

# Geometria
Lc = L + t/2
Ac = t * w
P = 2 * (t + w)
A_fin = 2 * (L * w + L * t) + w * t  # area superficial

# Parametro m
m = np.sqrt(h * P / (k_al * Ac))
mLc = m * Lc

# Eficiencia
eta = np.tanh(mLc) / mLc

print('Aleta Reta:')
print('  L = {:.0f} mm, t = {:.0f} mm, w = {:.0f} mm'.format(L*1000, t*1000, w*1000))
print('  Lc = {:.1f} mm'.format(Lc*1000))
print('  m = {:.2f} m^-1'.format(m))
print('  mLc = {:.3f}'.format(mLc))
print('  eta = {:.3f} ({:.1f}%)'.format(eta, eta*100))
print('  A_fin = {:.4f} m^2'.format(A_fin))
```

---

## 🌡️ Seção 2: Cálculo da Temperatura da Base

```python
# Calor por aleta
Q_por_aleta = Q_total / N_aletas

# theta_b (diferenca de temperatura base-ambiente)
theta_b = Q_por_aleta / (eta * h * A_fin)
T_base = T_inf + theta_b

# Temperatura do chip
T_chip = T_base + Q_total * R_chip_base

print('Resultados:')
print('  Q por aleta = {:.1f} W'.format(Q_por_aleta))
print('  theta_b = {:.1f} K'.format(theta_b))
print('  T_base = {:.1f} °C'.format(T_base))
print('  T_chip = {:.1f} °C'.format(T_chip))
print('  T_chip_max = {:.1f} °C'.format(T_chip_max))

if T_chip <= T_chip_max:
    print('\n✅ Temperatura DENTRO do limite')
else:
    print('\n❌ Temperatura EXCEDE o limite!')
    print('   Excesso: {:.1f} °C'.format(T_chip - T_chip_max))
```

---

## 📊 Seção 3: Análise de Sensibilidade

```python
# Variar numero de aletas
N_range = np.arange(4, 20)
T_chip_range = []

for N in N_range:
    Q_aleta = Q_total / N
    theta_b = Q_aleta / (eta * h * A_fin)
    T_b = T_inf + theta_b
    T_c = T_b + Q_total * R_chip_base
    T_chip_range.append(T_c)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(N_range, T_chip_range, 'bo-', linewidth=2, markersize=8)
ax.axhline(y=T_chip_max, color='r', linestyle='--', label='Limite: {} °C'.format(T_chip_max))
ax.axvline(x=N_aletas, color='g', linestyle=':', label='Projeto: {} aletas'.format(N_aletas))
ax.set_xlabel('Numero de Aletas', fontsize=12)
ax.set_ylabel('Temperatura do Chip [°C]', fontsize=12)
ax.set_title('Efeito do Numero de Aletas na Temperatura', fontsize=14, fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Encontrar numero minimo de aletas
N_min = N_range[np.array(T_chip_range) <= T_chip_max][0]
print('\nNumero minimo de aletas: {}'.format(N_min))
print('Projeto atual ({} aletas): {}'.format(N_aletas, 'OK' if N_aletas >= N_min else 'INSUFICIENTE'))
```

3. **Temperatura do chip:** ~{:.1f}°C (dentro do limite)
4. **Margem de segurança:** {:.1f}°C

### Recomendações
