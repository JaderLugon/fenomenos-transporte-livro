## 📋 Resumo

Simulamos a dispersão de um poluente conservativo lançado no Rio Macaé, comparando soluções analíticas e numéricas. O estudo inclui:

1. Cálculo dos coeficientes de dispersão (Liu, Elder)
2. Simulação de lançamento instantâneo e contínuo
3. Análise de sensibilidade aos parâmetros
4. Comparação com dados experimentais sintéticos

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erfc
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
print('Ambiente OK')
```

---

## 🌊 Seção 1: Parâmetros do Rio Macaé

```python
U = 0.20
B = 42.2
D = 0.71
S0 = 1e-4
g = 9.81
A = B * D

U_star = np.sqrt(g * D * S0)
E_L = 0.011 * (U**2 * B**2) / (D * U_star)
E_T = 0.23 * D * U_star

print('Parametros do Rio Macae:')
print('  U = {} m/s, B = {} m, D = {} m'.format(U, B, D))
print('  U* = {:.4f} m/s'.format(U_star))
print('  E_L = {:.2f} m2/s'.format(E_L))
print('  E_T = {:.4f} m2/s'.format(E_T))
```

---

## 🧪 Seção 2: Lançamento Instantâneo

```python
M = 10.0
tempos = [1800, 3600, 7200, 14400]

fig, ax = plt.subplots()
x = np.linspace(0, 5000, 500)

for t in tempos:
    C = M / (A * np.sqrt(4 * np.pi * E_L * t)) * np.exp(-(x - U * t)**2 / (4 * E_L * t))
    ax.plot(x/1000, C*1e6, lw=2, label='t = {} min'.format(t//60))

ax.set_xlabel('Distancia x [km]')
ax.set_ylabel('Concentracao [mg/L]')
ax.set_title('Evolucao da Pluma - Lancamento Instantaneo', fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

---

## 📈 Seção 3: Lançamento Contínuo

```python
C0 = 5.0
t_cont = 7200

termo1 = erfc((x - U * t_cont) / np.sqrt(4 * E_L * t_cont))
termo2 = np.exp(U * x / E_L) * erfc((x + U * t_cont) / np.sqrt(4 * E_L * t_cont))
C_cont = (C0 / 2) * (termo1 + termo2)

fig, ax = plt.subplots()
ax.plot(x/1000, C_cont, 'r-', lw=2)
ax.set_xlabel('Distancia x [km]')
ax.set_ylabel('Concentracao [mg/L]')
ax.set_title('Pluma - Lancamento Continuo (t = 2 h)', fontweight='bold')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

---

## 🔍 Seção 4: Análise de Sensibilidade

```python
E_L_range = np.linspace(5, 20, 10)
C_max_range = []

for E_L_t in E_L_range:
    C_max_t = M / (A * np.sqrt(4 * np.pi * E_L_t * t_cont))
    C_max_range.append(C_max_t)

fig, ax = plt.subplots()
ax.plot(E_L_range, np.array(C_max_range)*1e6, 'bo-', lw=2)
ax.set_xlabel('E_L [m2/s]')
ax.set_ylabel('C_max [mg/L]')
ax.set_title('Sensibilidade a E_L', fontweight='bold')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

---

## 💡 Conclusões

1. **Dispersão longitudinal** domina o espalhamento da pluma
2. **Advecção** transporta a pluma rio abaixo
3. **Sensibilidade** a $E_L$ é significativa (inversamente proporcional a $\sqrt{E_L}$)
4. **Monitoramento** deve considerar múltiplos pontos a jusante
