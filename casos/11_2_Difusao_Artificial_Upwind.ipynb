## 📋 Resumo

Analisamos a **difusão numérica artificial** introduzida pelo esquema upwind de primeira ordem e comparamos com esquemas de ordem superior (Crank-Nicolson, TVD).

**Objetivos:**
1. Implementar esquema upwind explícito
2. Quantificar difusão artificial
3. Comparar com solução analítica
4. Explorar esquemas de alta ordem

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
print('Ambiente OK')
```

---

## 🔧 Seção 1: Implementação do Esquema Upwind

```python
def upwind_explicito(C, u, E_L, dx, dt):
    C_new = C.copy()
    Co = u * dt / dx
    Di = E_L * dt / dx**2
    for i in range(1, len(C)-1):
        C_new[i] = C[i] - Co*(C[i] - C[i-1]) + Di*(C[i+1] - 2*C[i] + C[i-1])
    return C_new

print('Esquema upwind implementado')
```

---

## 📊 Seção 2: Simulação com Degrau

```python
N = 100
dx = 10.0
dt = 5.0
u = 0.5
E_L = 1.0

C = np.zeros(N)
C[N//2:] = 1.0

resultados = [C.copy()]
tempos = [0]

for step in range(500):
    C = upwind_explicito(C, u, E_L, dx, dt)
    if (step + 1) % 100 == 0:
        resultados.append(C.copy())
        tempos.append((step + 1) * dt)

fig, ax = plt.subplots()
x = np.arange(N) * dx
for t, C_res in zip(tempos, resultados):
    ax.plot(x, C_res, lw=2, label='t = {} s'.format(t))
ax.set_xlabel('Posicao x [m]')
ax.set_ylabel('Concentracao')
ax.set_title('Evolucao do Degrau - Esquema Upwind', fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

---

## 🔍 Seção 3: Quantificação da Difusão Artificial

```python
E_artificial = u * dx / 2
E_total = E_L + E_artificial

print('Difusao Artificial:')
print('  E_fisica = {:.2f} m2/s'.format(E_L))
print('  E_artificial = {:.2f} m2/s'.format(E_artificial))
print('  E_total = {:.2f} m2/s'.format(E_total))
print('  Contribuicao artificial: {:.1f}%'.format(E_artificial/E_total*100))
```

---

## 💡 Conclusões

1. **Upwind introduz difusão artificial** proporcional a $u \Delta x / 2$
2. **Malhas finas** reduzem o erro numérico
3. **Esquemas de alta ordem** (TVD, WENO) minimizam difusão artificial
4. **Sempre valide** resultados numéricos com soluções analíticas
