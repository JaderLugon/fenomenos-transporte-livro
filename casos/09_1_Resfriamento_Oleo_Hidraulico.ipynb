## 🏭 Seção 1: Descrição do Problema

### Contexto

Uma hidrelétrica de pequeno porte (5 MW) possui um sistema de lubrificação dos mancais da turbina que utiliza óleo hidráulico ISO VG 68. Durante a operação, o óleo absorve calor dos mancais e precisa ser resfriado antes de retornar ao reservatório.

### Dados do Sistema

| Parâmetro | Valor |
|-----------|-------|
| Vazão de óleo | $\dot{m}_h = 15$ kg/s |
| Temperatura de entrada do óleo | $T_{h,in} = 60$°C |
| Temperatura de saída desejada | $T_{h,out} = 45$°C |
| Calor específico do óleo | $c_{p,h} = 2000$ J/(kg·K) |
| Vazão de água de resfriamento | $\dot{m}_c = 12$ kg/s |
| Temperatura de entrada da água | $T_{c,in} = 20$°C |
| Calor específico da água | $c_{p,c} = 4180$ J/(kg·K) |
| Fonte de água | Rio (captação direta) |

```python
# ============================================================
# DADOS DO SISTEMA
# ============================================================

print("=" * 60)
print("DADOS DO SISTEMA DE RESFRIAMENTO")
print("=" * 60)

# Dados do óleo (fluido quente)
m_h = 15.0         # kg/s
cp_h = 2000.0      # J/(kg·K)
T_h_in = 60.0      # °C
T_h_out = 45.0     # °C

# Dados da água (fluido frio)
m_c = 12.0         # kg/s
cp_c = 4180.0      # J/(kg·K)
T_c_in = 20.0      # °C

# Cálculo da carga térmica
Q = m_h * cp_h * (T_h_in - T_h_out)

# Temperatura de saída da água (balanço de energia)
T_c_out = T_c_in + Q / (m_c * cp_c)

# Capacidades térmicas
C_h = m_h * cp_h
C_c = m_c * cp_c

print(f"\n📊 Dados do óleo:")
print(f"  • Vazão mássica: ṁ_h = {m_h} kg/s")
print(f"  • Calor específico: c_p,h = {cp_h} J/(kg·K)")
print(f"  • Temperatura entrada: T_h,in = {T_h_in}°C")
print(f"  • Temperatura saída: T_h,out = {T_h_out}°C")
print(f"  • Capacidade térmica: C_h = {C_h:.0f} W/K")

print(f"\n📊 Dados da água:")
print(f"  • Vazão mássica: ṁ_c = {m_c} kg/s")
print(f"  • Calor específico: c_p,c = {cp_c} J/(kg·K)")
print(f"  • Temperatura entrada: T_c,in = {T_c_in}°C")
print(f"  • Temperatura saída: T_c,out = {T_c_out:.2f}°C")
print(f"  • Capacidade térmica: C_c = {C_c:.0f} W/K")

print(f"\n🔋 Carga térmica:")
print(f"  • Q = ṁ_h · c_p,h · (T_h,in - T_h,out)")
print(f"  • Q = {m_h} × {cp_h} × ({T_h_in} - {T_h_out})")
print(f"  • Q = {Q:.0f} W = {Q/1000:.1f} kW")

print(f"\n💡 Interpretação:")
print(f"  A água tem maior capacidade térmica ({C_c:.0f} W/K vs {C_h:.0f} W/K),")
print(f"  por isso sofre menor variação de temperatura ({T_c_out - T_c_in:.1f}°C")
print(f"  vs {T_h_in - T_h_out:.1f}°C do óleo).")
```

## 🌡️ Seção 2: Cálculo da LMTD

### Trocador em Contracorrente

Para um trocador em contracorrente:

$$
\Delta T_1 = T_{h,in} - T_{c,out} \quad \text{(extremidade quente)}
$$

$$
\Delta T_2 = T_{h,out} - T_{c,in} \quad \text{(extremidade fria)}
$$

$$
\Delta T_{lm} = \frac{\Delta T_1 - \Delta T_2}{\ln(\Delta T_1 / \Delta T_2)}
$$

```python
# ============================================================
# CÁLCULO DA LMTD
# ============================================================

print("=" * 60)
print("CÁLCULO DA LMTD - CONTRACORRENTE")
print("=" * 60)

# Diferenças de temperatura
dT1 = T_h_in - T_c_out
dT2 = T_h_out - T_c_in

# LMTD
LMTD = (dT1 - dT2) / np.log(dT1 / dT2)

# Média aritmética (para comparação)
dT_arith = (dT1 + dT2) / 2

print(f"\n📊 Diferenças de temperatura:")
print(f"  • ΔT₁ = T_h,in - T_c,out = {T_h_in} - {T_c_out:.2f} = {dT1:.2f}°C")
print(f"  • ΔT₂ = T_h,out - T_c,in = {T_h_out} - {T_c_in} = {dT2:.2f}°C")
print(f"  • Razão ΔT₁/ΔT₂ = {dT1/dT2:.2f}")

print(f"\n🧮 LMTD:")
print(f"  • LMTD = (ΔT₁ - ΔT₂) / ln(ΔT₁/ΔT₂)")
print(f"  • LMTD = ({dT1:.2f} - {dT2:.2f}) / ln({dT1:.2f}/{dT2:.2f})")
print(f"  • LMTD = {LMTD:.2f}°C")

print(f"\n🔍 Comparação com média aritmética:")
print(f"  • Média aritmética: {dT_arith:.2f}°C")
print(f"  • LMTD: {LMTD:.2f}°C")
print(f"  • Erro: {abs(dT_arith - LMTD)/LMTD*100:.2f}%")

print(f"\n💡 Interpretação:")
print(f"  Como ΔT₁/ΔT₂ = {dT1/dT2:.2f} < 1,5, a média aritmética")
print(f"  seria aceitável, mas usamos LMTD por ser exata.")
```

## 🧱 Seção 3: Dimensionamento do Trocador

### Seleção do Tipo de Trocador

Para esta aplicação, escolhemos um **trocador de placas** devido a:

1. **Alta eficiência térmica**: coeficientes de transferência elevados
2. **Compactidade**: alta área por volume
3. **Facilidade de limpeza**: placas removíveis
4. **Flexibilidade**: número de placas ajustável

### Coeficiente Global Típico

Para óleo-água em trocador de placas:
$$
U \approx 600 \text{--} 1200 \text{ W/(m²·K)}
$$

Adotaremos $U = 800$ W/(m²·K) (valor conservador).

```python
# ============================================================
# DIMENSIONAMENTO DO TROCADOR
# ============================================================

print("=" * 60)
print("DIMENSIONAMENTO DO TROCADOR DE PLACAS")
print("=" * 60)

# Coeficiente global
U = 800.0        # W/(m²·K)
F = 0.95         # fator de correção (trocador de placas multipasse)

# Área necessária
A = Q / (U * F * LMTD)

# Dados do trocador de placas (fabricante típico)
A_placa = 1.2    # m² (área por placa)

# Número de placas
N_placas = np.ceil(A / A_placa)

# Área real instalada
A_instalada = N_placas * A_placa

# Margem de projeto
margem = (A_instalada - A) / A * 100

print(f"\n📊 Parâmetros de projeto:")
print(f"  • Coeficiente global: U = {U} W/(m²·K)")
print(f"  • Fator de correção: F = {F}")
print(f"  • LMTD = {LMTD:.2f}°C")

print(f"\n🧮 Área necessária:")
print(f"  • A = Q / (U · F · LMTD)")
print(f"  • A = {Q:.0f} / ({U} × {F} × {LMTD:.2f})")
print(f"  • A = {A:.2f} m²")

print(f"\n📐 Seleção do trocador:")
print(f"  • Área por placa: A_placa = {A_placa} m²")
print(f"  • Número de placas: N = ⌈{A:.2f}/{A_placa}⌉ = {N_placas:.0f}")
print(f"  • Área instalada: A_inst = {A_instalada:.2f} m²")
print(f"  • Margem de projeto: {margem:.1f}%")

print(f"\n💡 Interpretação:")
print(f"  A margem de {margem:.1f}% é adequada para compensar:")
print(f"    • Incrustação ao longo do tempo")
print(f"    • Variações nas condições operacionais")
print(f"    • Degradação do óleo")
```

---

## 📈 Seção 4: Verificação com Método ε-NTU

Vamos verificar o dimensionamento usando o método ε-NTU.

```python
# ============================================================
# VERIFICAÇÃO COM ε-NTU
# ============================================================

print("=" * 60)
print("VERIFICAÇÃO COM MÉTODO ε-NTU")
print("=" * 60)

# Capacidades térmicas
C_min = min(C_h, C_c)
C_max = max(C_h, C_c)
C_r = C_min / C_max

# NTU
NTU = U * A_instalada / C_min

# Efetividade (contracorrente)
exp_term = np.exp(-NTU * (1 - C_r))
epsilon = (1 - exp_term) / (1 - C_r * exp_term)

# Carga térmica máxima
Q_max = C_min * (T_h_in - T_c_in)
Q_calc = epsilon * Q_max

# Temperaturas de saída calculadas
T_h_out_calc = T_h_in - Q_calc / C_h
T_c_out_calc = T_c_in + Q_calc / C_c

print(f"\n📊 Parâmetros adimensionais:")
print(f"  • C_min = {C_min:.0f} W/K (óleo)")
print(f"  • C_max = {C_max:.0f} W/K (água)")
print(f"  • C_r = {C_r:.3f}")
print(f"  • NTU = {NTU:.3f}")

print(f"\n🧮 Resultados:")
print(f"  • ε = {epsilon:.3f}")
print(f"  • Q_max = {Q_max:.0f} W")
print(f"  • Q_calc = {Q_calc:.0f} W")
print(f"  • T_h,out (calc) = {T_h_out_calc:.2f}°C (espec: {T_h_out}°C)")
print(f"  • T_c,out (calc) = {T_c_out_calc:.2f}°C (espec: {T_c_out:.2f}°C)")

print(f"\n✅ Verificação:")
if abs(T_h_out_calc - T_h_out) < 1.0:
    print(f"  • Temperatura de saída do óleo dentro da especificação ✓")
else:
    print(f"  • Temperatura de saída do óleo FORA da especificação ✗")

print(f"\n💡 Interpretação:")
print(f"  A efetividade ε = {epsilon:.3f} indica que o trocador transfere")
print(f"  {epsilon*100:.1f}% do calor máximo possível.")
```

## 🔍 Seção 5: Análise de Incrustação

### Efeito da Incrustação no Desempenho

A incrustação (fouling) reduz o coeficiente global $U$ ao longo do tempo:

$$
\frac{1}{U_{\text{proj}}} = \frac{1}{U_{\text{limpo}}} + R_f
$$

onde $R_f$ é o fator de incrustação [m²·K/W].

### Fatores Típicos de Incrustação

| Fluido | $R_f$ [m²·K/W] |
|--------|----------------|
| Água de rio (limpa) | 0,0001 |
| Água de rio (poluída) | 0,0002–0,0005 |
| Óleo hidráulico | 0,0001 |
| Água do mar | 0,0001–0,0002 |

```python
# ============================================================
# ANÁLISE DE INCRUSTAÇÃO
# ============================================================

print("=" * 60)
print("ANÁLISE DE INCRUSTAÇÃO AO LONGO DO TEMPO")
print("=" * 60)

# Fatores de incrustação ao longo do tempo
tempo_meses = np.array([0, 3, 6, 9, 12])  # meses
Rf_valores = np.array([0, 0.0001, 0.0002, 0.0003, 0.0004])  # m²·K/W

# Calcular U_proj e desempenho para cada instante
U_limpo = U
U_proj_list = []
Q_proj_list = []
T_h_out_list = []

for Rf in Rf_valores:
    U_proj = 1 / (1/U_limpo + Rf)
    U_proj_list.append(U_proj)
    
    # Recalcular desempenho
    NTU_proj = U_proj * A_instalada / C_min
    exp_term_proj = np.exp(-NTU_proj * (1 - C_r))
    epsilon_proj = (1 - exp_term_proj) / (1 - C_r * exp_term_proj)
    Q_proj = epsilon_proj * Q_max
    Q_proj_list.append(Q_proj)
    
    T_h_out_proj = T_h_in - Q_proj / C_h
    T_h_out_list.append(T_h_out_proj)

U_proj_list = np.array(U_proj_list)
Q_proj_list = np.array(Q_proj_list)
T_h_out_list = np.array(T_h_out_list)

print(f"\n📊 Degradação do desempenho:")
print(f"{'Tempo [meses]':<15} {'R_f [m²·K/W]':<15} {'U_proj [W/(m²·K)]':<20} {'T_h,out [°C]':<12}")
print("-" * 62)
for i in range(len(tempo_meses)):
    print(f"{tempo_meses[i]:<15} {Rf_valores[i]:<15.4f} {U_proj_list[i]:<20.1f} {T_h_out_list[i]:<12.2f}")

print(f"\n🔍 Análise após 12 meses:")
print(f"  • U reduziu de {U_limpo:.0f} para {U_proj_list[-1]:.0f} W/(m²·K) ({(1 - U_proj_list[-1]/U_limpo)*100:.1f}%)")
print(f"  • T_h,out aumentou de {T_h_out:.1f}°C para {T_h_out_list[-1]:.1f}°C")

if T_h_out_list[-1] > 50:
    print(f"  • ⚠️ Temperatura de saída EXCEDE limite de 50°C")
    print(f"  • Necessária limpeza do trocador!")
else:
    print(f"  • ✓ Temperatura de saída dentro do limite")

# Plot
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Gráfico 1: U vs tempo
axes[0].plot(tempo_meses, U_proj_list, 'bo-', linewidth=2, markersize=8)
axes[0].set_xlabel('Tempo [meses]', fontsize=12)
axes[0].set_ylabel('Coeficiente Global U [W/(m²·K)]', fontsize=12)
axes[0].set_title('Degradação do Coeficiente Global', fontsize=13, fontweight='bold')
axes[0].grid(True, alpha=0.3)
axes[0].axhline(y=U_limpo, color='r', linestyle='--', label=f'U_limpo = {U_limpo:.0f}')
axes[0].legend(fontsize=10)

# Gráfico 2: T_h,out vs tempo
axes[1].plot(tempo_meses, T_h_out_list, 'ro-', linewidth=2, markersize=8)
axes[1].set_xlabel('Tempo [meses]', fontsize=12)
axes[1].set_ylabel('Temperatura de Saída do Óleo [°C]', fontsize=12)
axes[1].set_title('Aumento da Temperatura de Saída', fontsize=13, fontweight='bold')
axes[1].grid(True, alpha=0.3)
axes[1].axhline(y=50, color='orange', linestyle='--', linewidth=2, label='Limite = 50°C')
axes[1].axhline(y=T_h_out, color='g', linestyle=':', label=f'Especificação = {T_h_out}°C')
axes[1].legend(fontsize=10)

plt.tight_layout()
plt.show()
```

## 💰 Seção 6: Análise Econômica

### Custo da Perda de Eficiência

Vamos calcular o custo associado à degradação do trocador:

1. **Custo energético**: perda de eficiência térmica
2. **Custo de manutenção**: limpeza periódica
3. **Custo de parada**: perda de produção durante limpeza

```python
# ============================================================
# ANÁLISE ECONÔMICA
# ============================================================

print("=" * 60)
print("ANÁLISE ECONÔMICA")
print("=" * 60)

# Parâmetros econômicos
horas_operacao = 8000    # h/ano (operação contínua)
custo_limpeza = 5000     # R$ por limpeza
tempo_limpeza = 8        # horas por limpeza
custo_parada = 10000     # R$/hora (perda de produção)

# Calcular perda de eficiência ao longo do tempo
perda_eficiencia = (Q - Q_proj_list) / Q * 100  # %

print(f"\n📊 Parâmetros operacionais:")
print(f"  • Operação: {horas_operacao} h/ano")
print(f"  • Custo de limpeza: R$ {custo_limpeza:.0f}")
print(f"  • Tempo de limpeza: {tempo_limpeza} h")
print(f"  • Custo de parada: R$ {custo_parada:.0f}/h")

print(f"\n📈 Perda de eficiência ao longo do tempo:")
print(f"{'Tempo [meses]':<15} {'Perda [%]':<12} {'Q_perdido [kW]':<15}")
print("-" * 42)
for i in range(len(tempo_meses)):
    Q_perdido = (Q - Q_proj_list[i]) / 1000
    print(f"{tempo_meses[i]:<15} {perda_eficiencia[i]:<12.1f} {Q_perdido:<15.1f}")

# Custo anual de manutenção (limpeza a cada 6 meses)
limpezas_ano = 2
custo_manutencao_ano = limpezas_ano * (custo_limpeza + tempo_limpeza * custo_parada)

print(f"\n💰 Custo anual de manutenção:")
print(f"  • Limpezas por ano: {limpezas_ano}")
print(f"  • Custo por limpeza: R$ {custo_limpeza + tempo_limpeza*custo_parada:.0f}")
print(f"  • Custo anual: R$ {custo_manutencao_ano:.0f}")

print(f"\n💡 Recomendação:")
print(f"  • Limpeza a cada 6 meses para manter desempenho")
print(f"  • Monitorar T_h,out continuamente")
print(f"  • Se T_h,out > 48°C, programar limpeza imediata")
```

## 🌡️ Seção 7: Efeito da Temperatura da Água do Rio

### Variação Sazonal

A temperatura da água do rio varia ao longo do ano:
- Verão: 25–28°C
- Inverno: 18–20°C

Vamos analisar o efeito dessa variação no desempenho do trocador.

```python
# ============================================================
# EFEITO DA TEMPERATURA DA ÁGUA
# ============================================================

print("=" * 60)
print("EFEITO DA TEMPERATURA DA ÁGUA DO RIO")
print("=" * 60)

# Temperaturas da água ao longo do ano
T_c_in_range = np.linspace(18, 28, 11)  # °C

# Calcular desempenho para cada temperatura
T_h_out_sazonal = []
Q_sazonal = []

for T_c_in_teste in T_c_in_range:
    # Recalcular Q_max
    Q_max_teste = C_min * (T_h_in - T_c_in_teste)
    
    # Efetividade (mesma, pois NTU e C_r não mudam)
    Q_teste = epsilon * Q_max_teste
    Q_sazonal.append(Q_teste)
    
    T_h_out_teste = T_h_in - Q_teste / C_h
    T_h_out_sazonal.append(T_h_out_teste)

T_h_out_sazonal = np.array(T_h_out_sazonal)
Q_sazonal = np.array(Q_sazonal)

print(f"\n📊 Efeito da temperatura da água:")
print(f"{'T_c,in [°C]':<12} {'T_h,out [°C]':<12} {'Q [kW]':<12}")
print("-" * 36)
for i in range(len(T_c_in_range)):
    print(f"{T_c_in_range[i]:<12.1f} {T_h_out_sazonal[i]:<12.2f} {Q_sazonal[i]/1000:<12.1f}")

# Plot
fig, ax1 = plt.subplots(figsize=(10, 6))

color1 = 'tab:red'
ax1.set_xlabel('Temperatura da Água do Rio [°C]', fontsize=12)
ax1.set_ylabel('Temperatura de Saída do Óleo [°C]', fontsize=12, color=color1)
ax1.plot(T_c_in_range, T_h_out_sazonal, color=color1, linewidth=2, marker='o')
ax1.tick_params(axis='y', labelcolor=color1)
ax1.grid(True, alpha=0.3)

# Linha de limite
ax1.axhline(y=50, color='orange', linestyle='--', linewidth=2, label='Limite = 50°C')
ax1.axhline(y=T_h_out, color='g', linestyle=':', label=f'Especificação = {T_h_out}°C')
ax1.legend(fontsize=10, loc='upper left')

ax2 = ax1.twinx()
color2 = 'tab:blue'
ax2.set_ylabel('Carga Térmica [kW]', fontsize=12, color=color2)
ax2.plot(T_c_in_range, Q_sazonal/1000, color=color2, linewidth=2, marker='s', linestyle='--')
ax2.tick_params(axis='y', labelcolor=color2)

plt.title('Efeito Sazonal da Temperatura da Água', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print(f"\n💡 Interpretação:")
print(f"  • No verão (T_c,in = 28°C): T_h,out = {T_h_out_sazonal[-1]:.1f}°C")
print(f"  • No inverno (T_c,in = 18°C): T_h,out = {T_h_out_sazonal[0]:.1f}°C")
print(f"  • Variação: {T_h_out_sazonal[-1] - T_h_out_sazonal[0]:.1f}°C")
if T_h_out_sazonal[-1] > 50:
    print(f"  • ⚠️ No verão, T_h,out EXCEDE o limite!")
    print(f"  • Solução: aumentar vazão de água ou área de troca")
```

   - Troca de juntas a cada 3–5 anos

3. **Ações corretivas**:
   - Se $T_{h,out} > 48°C$: programar limpeza imediata
   - Se $T_{h,out} > 50°C$: reduzir carga ou aumentar vazão de água

### Limitações do Projeto

- **Variação sazonal**: no verão, com água a 28°C, o trocador pode não atender à especificação
- **Solução**: considerar aumento de área (25 placas) ou sistema de resfriamento prévio da água

### Aplicações Práticas

Este estudo de caso é aplicável a:
- Hidrelétricas de pequeno e médio porte
- Sistemas de lubrificação industrial
- Resfriamento de óleo em máquinas-ferramenta
- Sistemas hidráulicos de prensas e injetoras
