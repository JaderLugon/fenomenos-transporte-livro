## 🧱 Seção 1: Parede Residencial Típica

### Descrição

Vamos analisar uma parede residencial típica composta por:

| Camada | Espessura [cm] | k [W/(m·K)] |
|--------|----------------|-------------|
| Reboco interno | 2,0 | 0,8 |
| Tijolo baiano | 10,0 | 0,6 |
| Reboco externo | 2,0 | 0,8 |

**Condições:**
- $T_{int} = 22$°C (ar interno)
- $T_{ext} = 35$°C (ar externo)
- $h_{int} = 8$ W/(m²·K)
- $h_{ext} = 20$ W/(m²·K)

```python
# ============================================================
# PAREDE RESIDENCIAL TÍPICA
# ============================================================

print("=" * 60)
print("PAREDE RESIDENCIAL TÍPICA")
print("=" * 60)

# Dados
T_int = 22.0       # °C
T_ext = 35.0       # °C
h_int = 8.0        # W/(m²·K)
h_ext = 20.0       # W/(m²·K)

camadas = [
    {'nome': 'Reboco interno', 'L': 0.02, 'k': 0.8},
    {'nome': 'Tijolo baiano', 'L': 0.10, 'k': 0.6},
    {'nome': 'Reboco externo', 'L': 0.02, 'k': 0.8}
]

# Calcular resistências (por m²)
R_conv_int = 1 / h_int
R_conv_ext = 1 / h_ext

R_paredes = []
for camada in camadas:
    R = camada['L'] / camada['k']
    R_paredes.append(R)

R_total = R_conv_int + sum(R_paredes) + R_conv_ext

print(f"\n📊 Resistências térmicas (por m²):")
print(f"  • R_conv,interna = 1/h_int = {R_conv_int:.4f} m²·K/W")
for i, camada in enumerate(camadas):
    print(f"  • R_{camada['nome']} = {R_paredes[i]:.4f} m²·K/W")
print(f"  • R_conv,externa = 1/h_ext = {R_conv_ext:.4f} m²·K/W")
print(f"  • R_total = {R_total:.4f} m²·K/W")

# Fluxo de calor
q = (T_ext - T_int) / R_total

print(f"\n🔋 Fluxo de calor:")
print(f"  • q'' = (T_ext - T_int) / R_total")
print(f"  • q'' = ({T_ext} - {T_int}) / {R_total:.4f}")
print(f"  • q'' = {q:.1f} W/m²")

# Contribuição percentual
print(f"\n📈 Contribuição percentual:")
R_total_list = [R_conv_int] + R_paredes + [R_conv_ext]
nomes = ['Conv. interna'] + [c['nome'] for c in camadas] + ['Conv. externa']

for nome, R in zip(nomes, R_total_list):
    percentual = R / R_total * 100
    print(f"  • {nome:<20}: {percentual:5.1f}%")
```

## 🪟 Seção 2: Parede com Janela (Resistências em Paralelo)

### Descrição

Uma parede de 3 m × 2,5 m tem uma janela de 1,5 m × 1,2 m. Vamos calcular o fluxo total considerando as duas regiões em paralelo.

**Parede (opaca):** mesma composição do caso anterior  
**Janela:** vidro simples de 5 mm ($k = 1,0$ W/(m·K))

```python
# ============================================================
# PAREDE COM JANELA
# ============================================================

print("=" * 60)
print("PAREDE COM JANELA (RESISTÊNCIAS EM PARALELO)")
print("=" * 60)

# Dimensões
A_parede_total = 3.0 * 2.5      # m²
A_janela = 1.5 * 1.2            # m²
A_parede_opaca = A_parede_total - A_janela

# Janela (vidro simples)
L_vidro = 0.005  # m
k_vidro = 1.0    # W/(m·K)
R_janela = R_conv_int + L_vidro/k_vidro + R_conv_ext

# Fluxos
q_parede = (T_ext - T_int) / R_total * A_parede_opaca
q_janela = (T_ext - T_int) / R_janela * A_janela
q_total = q_parede + q_janela

print(f"\n📐 Áreas:")
print(f"  • Parede total: {A_parede_total:.2f} m²")
print(f"  • Janela: {A_janela:.2f} m²")
print(f"  • Parede opaca: {A_parede_opaca:.2f} m²")

print(f"\n🔋 Fluxos de calor:")
print(f"  • Pela parede opaca: {q_parede:.1f} W")
print(f"  • Pela janela: {q_janela:.1f} W")
print(f"  • Total: {q_total:.1f} W")

print(f"\n📊 Contribuição:")
print(f"  • Parede opaca: {q_parede/q_total*100:.1f}% da perda")
print(f"  • Janela: {q_janela/q_total*100:.1f}% da perda")
print(f"  • (Apesar de ser apenas {A_janela/A_parede_total*100:.1f}% da área!)")

# U global (coeficiente global de transmissão)
U_global = q_total / (A_parede_total * (T_ext - T_int))
print(f"\n📋 U global da parede com janela: {U_global:.2f} W/(m²·K)")
```

## 🔗 Seção 3: Efeito da Resistência de Contato

### Descrição

Quando duas camadas não estão perfeitamente aderidas, surge uma **resistência de contato** $R_c$ na interface. Vamos analisar o efeito desse fenômeno.

```python
# ============================================================
# RESISTÊNCIA DE CONTATO
# ============================================================

print("=" * 60)
print("EFEITO DA RESISTÊNCIA DE CONTATO")
print("=" * 60)

# Variar resistência de contato
Rc_values = [0, 0.01, 0.05, 0.10, 0.20]  # m²·K/W

print(f"\n📊 Efeito da resistência de contato entre tijolo e reboco:")
print(f"{'R_c [m²·K/W]':<18} {'R_total':<12} {'q'' [W/m²]':<12} {'Δq [%]':<10}")
print("-" * 55)

q_base = q  # fluxo sem resistência de contato

for Rc in Rc_values:
    R_total_Rc = R_total + Rc
    q_Rc = (T_ext - T_int) / R_total_Rc
    delta_q = (q_Rc - q_base) / q_base * 100
    print(f"{Rc:<18.2f} {R_total_Rc:<12.4f} {q_Rc:<12.1f} {delta_q:<+10.1f}")

print(f"\n💡 Interpretação:")
print(f"  • R_c típico para interfaces não-perfeitas: 0,01-0,1 m²·K/W")
print(f"  • Para paredes bem construídas: R_c ≈ 0 (adesão perfeita)")
print(f"  • Para paredes com falhas: R_c pode ser significativo")
print(f"  • Em metais (alta k), R_c é mais crítico que em isolantes")
```

## 📊 Seção 4: Comparação de Materiais

### Vamos comparar diferentes soluções de parede

```python
# ============================================================
# COMPARAÇÃO DE MATERIAIS
# ============================================================

print("=" * 60)
print("COMPARAÇÃO DE SOLUÇÕES DE PAREDE")
print("=" * 60)

solucoes = {
    'Parede simples (tijolo)': {
        'camadas': [{'L': 0.10, 'k': 0.6}],
        'custo_rel': 1.0
    },
    '+ Isolamento (EPS 3cm)': {
        'camadas': [{'L': 0.10, 'k': 0.6}, {'L': 0.03, 'k': 0.035}],
        'custo_rel': 1.5
    },
    '+ Isolamento (EPS 5cm)': {
        'camadas': [{'L': 0.10, 'k': 0.6}, {'L': 0.05, 'k': 0.035}],
        'custo_rel': 1.8
    },
    'Parede dupla (ar 5cm)': {
        'camadas': [{'L': 0.10, 'k': 0.6}, {'L': 0.05, 'k': 0.026}, {'L': 0.10, 'k': 0.6}],
        'custo_rel': 2.5
    },
    'Parede de concreto (15cm)': {
        'camadas': [{'L': 0.15, 'k': 1.7}],
        'custo_rel': 1.3
    }
}

print(f"\n{'Solução':<30} {'R_total':<12} {'q'' [W/m²]':<12} {'Custo rel.':<12}")
print("-" * 70)

resultados = []
for nome, dados in solucoes.items():
    R_paredes_sol = sum(c['L']/c['k'] for c in dados['camadas'])
    R_total_sol = R_conv_int + R_paredes_sol + R_conv_ext
    q_sol = (T_ext - T_int) / R_total_sol
    resultados.append((nome, R_total_sol, q_sol, dados['custo_rel']))
    print(f"{nome:<30} {R_total_sol:<12.4f} {q_sol:<12.1f} {dados['custo_rel']:<12.1f}")

# Plot comparativo
nomes = [r[0] for r in resultados]
fluxos = [r[2] for r in resultados]
custos = [r[3] for r in resultados]

fig, ax = plt.subplots(figsize=(12, 6))

x = np.arange(len(nomes))
width = 0.4

bars1 = ax.bar(x - width/2, fluxos, width, label='Fluxo de calor [W/m²]', color='red', alpha=0.7)
ax2 = ax.twinx()
bars2 = ax2.bar(x + width/2, custos, width, label='Custo relativo', color='blue', alpha=0.7)

ax.set_xlabel('Solução de Parede', fontsize=12)
ax.set_ylabel('Fluxo de Calor [W/m²]', fontsize=12, color='red')
ax2.set_ylabel('Custo Relativo', fontsize=12, color='blue')
ax.set_title('Comparação de Soluções de Parede', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(nomes, rotation=15, ha='right', fontsize=9)
ax.tick_params(axis='y', labelcolor='red')
ax2.tick_params(axis='y', labelcolor='blue')

# Legenda combinada
lines1, labels1 = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines1 + lines2, labels1 + labels2, loc='upper right', fontsize=10)

plt.tight_layout()
plt.show()

print(f"\n💡 Interpretação:")
print(f"  • Menor fluxo = melhor isolamento térmico")
print(f"  • Parede dupla com ar é a melhor solução (menor fluxo)")
print(f"  • Concreto tem pior desempenho (alta k)")
print(f"  • O custo-benefício depende do clima local e do uso")
```

## 💡 Seção 5: Discussão e Conclusões

### Conceitos-Chave

1. **Resistências em série:** Somam-se diretamente
   $$R_{\text{total}} = R_1 + R_2 + R_3 + \cdots$$

2. **Resistências em paralelo:** Inverso da soma dos inversos
   $$\frac{1}{R_{\text{total}}} = \frac{1}{R_1} + \frac{1}{R_2} + \cdots$$

3. **Camada dominante:** A camada com maior resistência controla o fluxo total

4. **Resistência de contato:** Pode ser significativa em interfaces mal-executadas

### Aplicações Práticas

- **Arquitetura:** Dimensionamento de isolamento de edificações
- **Refrigeração:** Câmaras frigoríficas, freezers
- **Indústria:** Fornos, caldeiras, trocadores de calor
- **Conforto térmico:** Análise de paredes para certificações (LEED, AQUA)
