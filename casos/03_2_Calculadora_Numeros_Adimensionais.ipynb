## 🧮 Funções de Cálculo

Vamos implementar as funções para calcular os principais números adimensionais.

```
# ============================================================
# FUNÇÕES DE CÁLCULO
# ============================================================

def calcular_reynolds(U, L, nu):
    """
    Calcula o número de Reynolds.
    
    Re = U * L / ν
    
    Parâmetros:
    -----------
    U : float
        Velocidade característica [m/s]
    L : float
        Comprimento característico [m]
    nu : float
        Viscosidade cinemática [m²/s]
    
    Retorna:
    --------
    float
        Número de Reynolds [adimensional]
    """
    return U * L / nu

def calcular_peclet(U, L, D):
    """
    Calcula o número de Péclet.
    
    Pe = U * L / D
    
    Parâmetros:
    -----------
    U : float
        Velocidade característica [m/s]
    L : float
        Comprimento característico [m]
    D : float
        Difusividade (térmica α, mássica D, ou dispersão E) [m²/s]
    
    Retorna:
    --------
    float
        Número de Péclet [adimensional]
    """
    return U * L / D

def calcular_prandtl(nu, alpha):
    """
    Calcula o número de Prandtl.
    
    Pr = ν / α
    
    Parâmetros:
    -----------
    nu : float
        Viscosidade cinemática [m²/s]
    alpha : float
        Difusividade térmica [m²/s]
    
    Retorna:
    --------
    float
        Número de Prandtl [adimensional]
    """
    return nu / alpha

def calcular_schmidt(nu, D):
    """
    Calcula o número de Schmidt.
    
    Sc = ν / D
    
    Parâmetros:
    -----------
    nu : float
        Viscosidade cinemática [m²/s]
    D : float
        Difusividade mássica [m²/s]
    
    Retorna:
    --------
    float
        Número de Schmidt [adimensional]
    """
    return nu / D

def calcular_froude(U, L, g=9.81):
    """
    Calcula o número de Froude.
    
    Fr = U / √(g * L)
    
    Parâmetros:
    -----------
    U : float
        Velocidade característica [m/s]
    L : float
        Comprimento característico [m]
    g : float, opcional
        Aceleração da gravidade [m/s²] (padrão: 9,81)
    
    Retorna:
    --------
    float
        Número de Froude [adimensional]
    """
    return U / np.sqrt(g * L)

def calcular_damkohler(k_rxn, L, U):
    """
    Calcula o número de Damköhler.
    
    Da = k_rxn * L / U
    
    Parâmetros:
    -----------
    k_rxn : float
        Taxa de reação [1/s]
    L : float
        Comprimento característico [m]
    U : float
        Velocidade característica [m/s]
    
    Retorna:
    --------
    float
        Número de Damköhler [adimensional]
    """
    return k_rxn * L / U

def classificar_regime(Re, Re_lam=2000, Re_turb=2400):
    """Classifica o regime de escoamento"""
    if Re < Re_lam:
        return "LAMINAR"
    elif Re > Re_turb:
        return "TURBULENTO"
    else:
        return "TRANSIÇÃO"

def classificar_transporte(Pe):
    """Classifica o regime de transporte"""
    if Pe > 10:
        return "ADVECÇÃO DOMINANTE"
    elif Pe < 0.1:
        return "DIFUSÃO DOMINANTE"
    else:
        return "MISTO"

def classificar_froude(Fr):
    """Classifica o regime de superfície livre"""
    if Fr < 0.95:
        return "SUBCRÍTICO (fluvial)"
    elif Fr > 1.05:
        return "SUPERCRÍTICO (torrencial)"
    else:
        return "CRÍTICO"

print("✅ Funções de cálculo implementadas!")
```

## 🌊 Cenários de Engenharia

Vamos calcular os números adimensionais para diferentes cenários.

```
# ============================================================
# CENÁRIO 1: TUBULAÇÃO RESIDENCIAL
# ============================================================

print("=" * 60)
print("CENÁRIO 1: TUBULAÇÃO RESIDENCIAL")
print("=" * 60)

# Dados
U_tubo = 1.0        # m/s
D_tubo = 0.025      # m (25 mm)
nu_agua = 1.0e-6    # m²/s
alpha_agua = 1.4e-7 # m²/s
D_sal = 1.5e-9      # m²/s

# Cálculos
Re_tubo = calcular_reynolds(U_tubo, D_tubo, nu_agua)
Pe_termico = calcular_peclet(U_tubo, D_tubo, alpha_agua)
Pe_massico = calcular_peclet(U_tubo, D_tubo, D_sal)
Pr = calcular_prandtl(nu_agua, alpha_agua)
Sc = calcular_schmidt(nu_agua, D_sal)

print(f"\n📊 Parâmetros:")
print(f"  • Velocidade: U = {U_tubo} m/s")
print(f"  • Diâmetro: D = {D_tubo*1000} mm")
print(f"  • Fluido: Água a 20°C")

print(f"\n🔢 Números Adimensionais:")
print(f"  • Re = {Re_tubo:.0f} → {classificar_regime(Re_tubo)}")
print(f"  • Pe_térmico = {Pe_termico:.0f} → {classificar_transporte(Pe_termico)}")
print(f"  • Pe_mássico = {Pe_massico:.0f} → {classificar_transporte(Pe_massico)}")
print(f"  • Pr = {Pr:.2f}")
print(f"  • Sc = {Sc:.0f}")

print(f"\n💡 Interpretação:")
print(f"  • Escoamento TURBULENTO (Re >> 2400)")
print(f"  • Transporte dominado por ADVECÇÃO (Pe >> 10)")
print(f"  • Pr > 1 → camada limite térmica mais fina que a hidrodinâmica")
print(f"  • Sc >> 1 → sal difunde muito mais devagar que o momentum")
```

```
# ============================================================
# CENÁRIO 2: RIO MACAÉ
# ============================================================

print("\n" + "=" * 60)
print("CENÁRIO 2: RIO MACAÉ (RJ)")
print("=" * 60)

# Dados
U_rio = 0.20        # m/s
L_rio = 42.2        # m (largura)
D_rio = 0.71        # m (profundidade)
E_L = 12.57         # m²/s (dispersão longitudinal)
E_T = 0.00431       # m²/s (dispersão transversal)

# Cálculos
Re_rio = calcular_reynolds(U_rio, D_rio, nu_agua)
Pe_L = calcular_peclet(U_rio, L_rio, E_L)
Pe_T = calcular_peclet(U_rio, L_rio, E_T)
Fr_rio = calcular_froude(U_rio, D_rio)

print(f"\n📊 Parâmetros:")
print(f"  • Velocidade: U = {U_rio} m/s")
print(f"  • Largura: B = {L_rio} m")
print(f"  • Profundidade: D = {D_rio} m")
print(f"  • E_L = {E_L} m²/s | E_T = {E_T} m²/s")

print(f"\n🔢 Números Adimensionais:")
print(f"  • Re = {Re_rio:.0f} → {classificar_regime(Re_rio)}")
print(f"  • Pe_L (longitudinal) = {Pe_L:.2f} → {classificar_transporte(Pe_L)}")
print(f"  • Pe_T (transversal) = {Pe_T:.0f} → {classificar_transporte(Pe_T)}")
print(f"  • Fr = {Fr_rio:.3f} → {classificar_froude(Fr_rio)}")

print(f"\n💡 Interpretação:")
print(f"  • Escoamento TURBULENTO (Re >> 2400)")
print(f"  • Pe_L ≈ 1 → advecção e dispersão longitudinal comparáveis")
print(f"  • Pe_T >> 1 → dispersão transversal muito lenta")
print(f"  • Fr < 1 → regime FLUVIAL (ondas podem subir o rio)")
```

```
# ============================================================
# CENÁRIO 3: ATMOSFERA (VENTO EM EDIFÍCIO)
# ============================================================

print("\n" + "=" * 60)
print("CENÁRIO 3: VENTO EM EDIFÍCIO")
print("=" * 60)

# Dados
U_vento = 10.0       # m/s (vento moderado)
L_edif = 30.0        # m (altura do edifício)
nu_ar = 1.5e-5       # m²/s (ar a 20°C)
alpha_ar = 2.2e-5    # m²/s
D_umidade = 2.4e-5   # m²/s (difusividade do vapor d'água)

# Cálculos
Re_edif = calcular_reynolds(U_vento, L_edif, nu_ar)
Pe_termico_edif = calcular_peclet(U_vento, L_edif, alpha_ar)
Pe_massico_edif = calcular_peclet(U_vento, L_edif, D_umidade)
Pr_ar = calcular_prandtl(nu_ar, alpha_ar)
Sc_ar = calcular_schmidt(nu_ar, D_umidade)

print(f"\n📊 Parâmetros:")
print(f"  • Velocidade do vento: U = {U_vento} m/s")
print(f"  • Altura do edifício: L = {L_edif} m")
print(f"  • Fluido: Ar a 20°C")

print(f"\n🔢 Números Adimensionais:")
print(f"  • Re = {Re_edif:.0e} → {classificar_regime(Re_edif)}")
print(f"  • Pe_térmico = {Pe_termico_edif:.0e} → {classificar_transporte(Pe_termico_edif)}")
print(f"  • Pe_mássico = {Pe_massico_edif:.0e} → {classificar_transporte(Pe_massico_edif)}")
print(f"  • Pr = {Pr_ar:.2f}")
print(f"  • Sc = {Sc_ar:.2f}")

print(f"\n💡 Interpretação:")
print(f"  • Escoamento ALTAMENTE TURBULENTO (Re ~ 10⁷)")
print(f"  • Transporte dominado por ADVECÇÃO (Pe ~ 10⁷)")
print(f"  • Pr ≈ 1 → camadas limite térmica e hidrodinâmica similares")
print(f"  • Sc ≈ 1 → transporte de umidade e momentum em escalas similares")
```

## 💡 Discussão

### Importância dos Números Adimensionais

1. **Classificação de Regimes:** Re determina se o escoamento é laminar ou turbulento, o que afeta drasticamente as correlações de transferência de calor e massa.

2. **Simplificação de Modelos:** Pe indica se podemos desprezar difusão (Pe >> 1) ou advecção (Pe << 1), simplificando as equações governantes.

3. **Similaridade:** Dois sistemas com mesmos números adimensionais têm comportamento físico similar, permitindo usar modelos em escala reduzida.

4. **Análise de Sensibilidade:** Números adimensionais agrupam parâmetros, reduzindo o número de variáveis em análises de sensibilidade.

### Aplicações Práticas

- **Re:** Dimensionamento de tubulações, seleção de bombas, previsão de perda de carga
- **Pe:** Modelagem de dispersão de poluentes, projeto de reatores químicos
- **Pr/Sc:** Projeto de trocadores de calor, análise de secagem, modelagem de evaporação
- **Fr:** Projeto de vertedouros, análise de ressaltos hidráulicos, modelagem de estuários
- **Da:** Projeto de reatores químicos, modelagem de degradação de poluentes
