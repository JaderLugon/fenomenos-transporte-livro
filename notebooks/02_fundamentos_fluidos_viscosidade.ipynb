## 📚 Conteúdo Teórico

### 2.1 Estática dos Fluidos

A pressão em um fluido em repouso varia com a profundidade segundo:

$$p = p_0 + \rho g h$$

onde:
- $p_0$ = pressão na superfície [Pa]
- $\rho$ = densidade do fluido [kg/m³]
- $g$ = aceleração da gravidade [m/s²]
- $h$ = profundidade [m]

### 2.2 Lei da Viscosidade de Newton

Para fluidos newtonianos, a tensão de cisalhamento é proporcional ao gradiente de velocidade:

$$\tau = \mu \frac{du}{dy}$$

onde $\mu$ é a viscosidade dinâmica [Pa·s].

### 2.3 Viscosidade Cinemática

$$\nu = \frac{\mu}{\rho} \quad \text{[m²/s]}$$

## 🔬 Seção 1: Estática dos Fluidos

### Exercício 1.1: Pressão Absoluta no Fundo de um Tanque

Calcule a pressão absoluta no fundo de um tanque de água com 3 m de profundidade.

```
# ============================================================
# EXERCÍCIO 1.1: Pressão Absoluta
# ============================================================

print("=" * 60)
print("EXERCÍCIO 1.1: PRESSÃO ABSOLUTA NO FUNDO DE UM TANQUE")
print("=" * 60)

# Dados do problema
p_atm = 100e3  # Pa (pressão atmosférica)
rho = 1000     # kg/m³ (densidade da água)
g = 9.81       # m/s²
h = 3.0        # m (profundidade)

# Cálculo
p_man = rho * g * h  # Pressão manométrica
p_abs = p_atm + p_man  # Pressão absoluta

print(f"\n📊 Dados do problema:")
print(f"  • Pressão atmosférica: p_atm = {p_atm/1000:.1f} kPa")
print(f"  • Densidade da água: ρ = {rho} kg/m³")
print(f"  • Profundidade: h = {h} m")

print(f"\n🧮 Cálculo:")
print(f"  • Pressão manométrica: p_man = ρ·g·h = {rho} × {g} × {h}")
print(f"                        = {p_man:.0f} Pa = {p_man/1000:.2f} kPa")
print(f"  • Pressão absoluta: p_abs = p_atm + p_man")
print(f"                    = {p_atm/1000:.1f} + {p_man/1000:.2f}")
print(f"                    = {p_abs/1000:.2f} kPa")

print(f"\n💡 Interpretação:")
print(f"  A pressão hidrostática contribui com {p_man/p_abs*100:.1f}% do valor absoluto.")
print(f"  Isso mostra que mesmo em pequenas profundidades (3 m),")
print(f"  a coluna d'água representa parcela significativa da pressão total.")
```

### Exercício 1.2: Manômetro em U

Calcule a pressão manométrica medida por um manômetro em U com óleo.

```
# ============================================================
# EXERCÍCIO 1.2: Manômetro em U
# ============================================================

print("\n" + "=" * 60)
print("EXERCÍCIO 1.2: MANÔMETRO EM U COM ÓLEO")
print("=" * 60)

# Dados
rho_oleo = 850   # kg/m³
h_coluna = 0.25  # m (25 cm)

# Cálculo
p_man = rho_oleo * g * h_coluna

print(f"\n📊 Dados do problema:")
print(f"  • Densidade do óleo: ρ_óleo = {rho_oleo} kg/m³")
print(f"  • Diferença de altura: h = {h_coluna*100:.0f} cm")

print(f"\n🧮 Cálculo:")
print(f"  • p_man = ρ_óleo · g · h = {rho_oleo} × {g} × {h_coluna}")
print(f"          = {p_man:.1f} Pa = {p_man/1000:.2f} kPa")

print(f"\n💡 Interpretação:")
print(f"  A baixa densidade do óleo (850 kg/m³) em relação ao mercúrio")
print(f"  (13.600 kg/m³) torna esse manômetro mais sensível para medir")
print(f"  pequenas diferenças de pressão.")
```

## 🧪 Seção 2: Viscosidade e Lei de Newton

### Exercício 2.1: Força Tangencial em Placa Móvel

Calcule a força necessária para puxar uma placa sobre uma camada de óleo.

```
# ============================================================
# EXERCÍCIO 2.1: Placa Móvel sobre Óleo
# ============================================================

print("\n" + "=" * 60)
print("EXERCÍCIO 2.1: FORÇA TANGENCIAL EM PLACA MÓVEL")
print("=" * 60)

# Dados
A = 0.5      # m² (área da placa)
h_oleo = 0.002  # m (2 mm de espessura)
U = 0.8      # m/s (velocidade da placa)
mu = 0.04    # Pa·s (viscosidade do óleo)

# Cálculo
du_dy = U / h_oleo  # Gradiente de velocidade
tau = mu * du_dy     # Tensão de cisalhamento
F = tau * A          # Força tangencial

print(f"\n📊 Dados do problema:")
print(f"  • Área da placa: A = {A} m²")
print(f"  • Espessura da camada: h = {h_oleo*1000:.0f} mm")
print(f"  • Velocidade: U = {U} m/s")
print(f"  • Viscosidade dinâmica: μ = {mu} Pa·s")

print(f"\n🧮 Cálculo:")
print(f"  • Gradiente de velocidade: du/dy = U/h = {U}/{h_oleo} = {du_dy:.0f} s⁻¹")
print(f"  • Tensão de cisalhamento: τ = μ·(du/dy) = {mu} × {du_dy:.0f} = {tau:.1f} Pa")
print(f"  • Força tangencial: F = τ·A = {tau:.1f} × {A} = {F:.1f} N")

print(f"\n💡 Interpretação:")
print(f"  O escoamento entre as placas é o escoamento de Couette,")
print(f"  com perfil de velocidade linear. A força de {F:.1f} N representa")
print(f"  o atrito viscoso que deve ser vencido para manter a placa em movimento.")
```

### Exercício 2.2: Viscosidade Cinemática

Converta viscosidade dinâmica em cinemática e entre unidades.

```
# ============================================================
# EXERCÍCIO 2.2: Viscosidade Cinemática
# ============================================================

print("\n" + "=" * 60)
print("EXERCÍCIO 2.2: VISCOSIDADE CINEMÁTICA")
print("=" * 60)

# Dados
mu = 0.15    # Pa·s
rho = 850    # kg/m³

# Cálculo
nu = mu / rho  # m²/s
nu_cSt = nu * 1e6  # centistokes (1 cSt = 10⁻⁶ m²/s)

print(f"\n📊 Dados do problema:")
print(f"  • Viscosidade dinâmica: μ = {mu} Pa·s")
print(f"  • Densidade: ρ = {rho} kg/m³")

print(f"\n🧮 Cálculo:")
print(f"  • ν = μ/ρ = {mu}/{rho} = {nu:.2e} m²/s")
print(f"  • ν = {nu_cSt:.1f} cSt (centistokes)")

print(f"\n💡 Interpretação:")
print(f"  Esse valor é típico de óleos lubrificantes leves em temperatura ambiente.")
print(f"  Para comparação:")
print(f"    • Água a 20°C: ν ≈ 1,0 cSt")
print(f"    • Óleo SAE 30: ν ≈ 350 cSt")
```

## 📈 Seção 4: Visualização Comparativa

Vamos visualizar o comportamento de diferentes tipos de fluidos.

```
# ============================================================
# SEÇÃO 4: VISUALIZAÇÃO DE FLUIDOS
# ============================================================

# Dados para os diferentes fluidos
gamma_dot = np.linspace(0, 100, 100)  # Taxa de deformação [s⁻¹]

# Newtoniano: τ = μ × γ̇ com μ = 0.1 Pa·s
tau_newtoniano = 0.1 * gamma_dot

# Pseudoplástico (Lei de Potência): τ = K × γ̇^n com K=0.5, n=0.6
tau_pseudoplastico = 0.5 * gamma_dot**0.6

# Dilatante (Lei de Potência): τ = K × γ̇^n com K=0.01, n=1.5
tau_dilatante = 0.01 * gamma_dot**1.5

# Plástico de Bingham: τ = τ_0 + μ_p × γ̇ com τ_0=5 Pa, μ_p=0.08 Pa·s
tau_bingham = 5 + 0.08 * gamma_dot
tau_bingham[gamma_dot == 0] = 0  # Não escoa abaixo de τ_0

# Plot
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(gamma_dot, tau_newtoniano, 'b-', linewidth=2, label='Newtoniano (μ = 0.1 Pa·s)')
ax.plot(gamma_dot, tau_pseudoplastico, 'r-', linewidth=2, label='Pseudoplástico (n = 0.6)')
ax.plot(gamma_dot, tau_dilatante, 'g-', linewidth=2, label='Dilatante (n = 1.5)')
ax.plot(gamma_dot, tau_bingham, 'm-', linewidth=2, label='Plástico de Bingham (τ₀ = 5 Pa)')

ax.set_xlabel('Taxa de Deformação, $\\dot{\\gamma}$ [s⁻¹]', fontsize=12)
ax.set_ylabel('Tensão de Cisalhamento, $\\tau$ [Pa]', fontsize=12)
ax.set_title('Comportamento Reológico de Diferentes Fluidos', fontsize=14, fontweight='bold')
ax.legend(fontsize=11, loc='upper left')
ax.grid(True, alpha=0.3)
ax.set_xlim([0, 100])
ax.set_ylim([0, 20])

plt.tight_layout()
plt.show()

print("\n💡 Observe que:")
print("   • Newtoniano: reta passando pela origem")
print("   • Pseudoplástico: curva côncava para baixo (viscosidade diminui)")
print("   • Dilatante: curva côncava para cima (viscosidade aumenta)")
print("   • Bingham: reta não passando pela origem (tensão de escoamento)")
```
