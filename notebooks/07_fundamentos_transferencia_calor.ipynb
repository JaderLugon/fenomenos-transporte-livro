## 📚 Conteúdo Teórico Resumido

### 7.1 Calor vs. Temperatura

- **Temperatura ($T$)**: propriedade intensiva que mede a agitação molecular [K ou °C]
- **Calor ($Q$)**: energia em trânsito devido a diferença de temperatura [J]

### 7.2 Calor Sensível e Latente

**Calor sensível** (varia a temperatura):
$$
Q = m \cdot c_p \cdot \Delta T
$$

**Calor latente** (mudança de fase):
$$
Q = m \cdot L
$$

### 7.3 Três Mecanismos de Transferência

| Mecanismo | Lei | Fórmula |
|-----------|-----|----------|
| Condução | Fourier | $q'' = -k \frac{dT}{dx}$ |
| Convecção | Newton | $q'' = h(T_s - T_\infty)$ |
| Radiação | Stefan-Boltzmann | $q'' = \varepsilon \sigma (T_s^4 - T_{\text{viz}}^4)$ |

### 7.4 Resistência Térmica

**Condução em parede plana:**
$$
R_{\text{cond}} = \frac{L}{k \cdot A} \quad \text{[K/W]}
$$

**Convecção:**
$$
R_{\text{conv}} = \frac{1}{h \cdot A} \quad \text{[K/W]}
$$

**Radiação (linearizada):**
$$
R_{\text{rad}} = \frac{1}{h_{\text{rad}} \cdot A}, \quad h_{\text{rad}} = \varepsilon \sigma (T_s + T_{\text{viz}})(T_s^2 + T_{\text{viz}}^2)
$$

### 7.5 Analogia Elétrica

$$
\dot{Q} = \frac{\Delta T}{R_{\text{total}}} \quad \leftrightarrow \quad I = \frac{\Delta V}{R}
$$

## 🔥 Seção 1: Calor Sensível e Equilíbrio Térmico

### Exercício 1.1: Aquecimento de Água

Calcule o calor necessário para aquecer 2 kg de água de 20°C para 80°C. Use $c_p = 4180$ J/(kg·K).

```python
# ============================================================
# EXERCÍCIO 1.1: Calor Sensível
# ============================================================

print("=" * 60)
print("EXERCÍCIO 1.1: AQUECIMENTO DE ÁGUA")
print("=" * 60)

# Dados
m = 2.0         # kg
T_i = 20.0      # °C
T_f = 80.0      # °C
c_p = 4180.0    # J/(kg·K)

# Cálculo
delta_T = T_f - T_i
Q = m * c_p * delta_T

print(f"\n📊 Dados do problema:")
print(f"  • Massa: m = {m} kg")
print(f"  • Temperatura inicial: T_i = {T_i}°C")
print(f"  • Temperatura final: T_f = {T_f}°C")
print(f"  • Calor específico: c_p = {c_p} J/(kg·K)")

print(f"\n🧮 Cálculo:")
print(f"  • ΔT = {T_f} - {T_i} = {delta_T} K")
print(f"  • Q = m·c_p·ΔT = {m} × {c_p} × {delta_T}")
print(f"  • Q = {Q:.0f} J = {Q/1000:.2f} kJ")
print(f"  • Q = {Q/4184:.2f} kcal")

print(f"\n💡 Interpretação:")
print(f"  O elevado calor específico da água (4180 J/(kg·K))")
print(f"  explica por que ela é amplamente usada em sistemas")
print(f"  de refrigeração e aquecimento: armazena muito calor")
print(f"  com pouca variação de temperatura.")
```

### Exercício 1.2: Equilíbrio Térmico Cobre-Água

Um bloco de cobre de 500 g a 150°C é colocado em 1 kg de água a 20°C. Calcule a temperatura final de equilíbrio.

```python
# ============================================================
# EXERCÍCIO 1.2: Equilíbrio Térmico
# ============================================================

print("=" * 60)
print("EXERCÍCIO 1.2: EQUILÍBRIO TÉRMICO COBRE-ÁGUA")
print("=" * 60)

# Dados
m_Cu = 0.5          # kg
T_Cu = 150.0        # °C
c_p_Cu = 385.0      # J/(kg·K)

m_agua = 1.0        # kg
T_agua = 20.0       # °C
c_p_agua = 4180.0   # J/(kg·K)

# Cálculo da temperatura final (conservação de energia)
# Q_Cu + Q_agua = 0
# m_Cu·c_p_Cu·(T_f - T_Cu) + m_agua·c_p_agua·(T_f - T_agua) = 0

T_f = (m_Cu * c_p_Cu * T_Cu + m_agua * c_p_agua * T_agua) / \
      (m_Cu * c_p_Cu + m_agua * c_p_agua)

print(f"\n📊 Dados do problema:")
print(f"  • Cobre: m = {m_Cu} kg, T = {T_Cu}°C, c_p = {c_p_Cu} J/(kg·K)")
print(f"  • Água: m = {m_agua} kg, T = {T_agua}°C, c_p = {c_p_agua} J/(kg·K)")

print(f"\n🧮 Cálculo:")
print(f"  • T_f = (m_Cu·c_p_Cu·T_Cu + m_agua·c_p_agua·T_agua) / ")
print(f"         (m_Cu·c_p_Cu + m_agua·c_p_agua)")
print(f"  • T_f = {T_f:.2f}°C")

print(f"\n💡 Interpretação:")
print(f"  A temperatura final ({T_f:.1f}°C) está muito próxima da")
print(f"  temperatura inicial da água (20°C), pois a água tem")
print(f"  c_p ~11× maior que o cobre e há o dobro de massa.")
```

## 🔄 Seção 2: Conversão de Unidades Térmicas

### Exercício 2.1: Conversor de Unidades

Converta: (a) 500 kcal para kJ; (b) 3 kW para BTU/h; (c) 200 W/m² para kcal/(h·m²).

```python
# ============================================================
# EXERCÍCIO 2.1: Conversão de Unidades
# ============================================================

print("=" * 60)
print("CONVERSOR DE UNIDADES TÉRMICAS")
print("=" * 60)

# Fatores de conversão
kcal_para_kJ = 4.184
W_para_BTUh = 3.413
Wm2_para_kcalhm2 = 0.8598

# (a) 500 kcal → kJ
val_a = 500 * kcal_para_kJ
print(f"\n(a) 500 kcal = {val_a:.1f} kJ")

# (b) 3 kW → BTU/h
val_b = 3 * 1000 * W_para_BTUh
print(f"(b) 3 kW = {val_b:.0f} BTU/h")

# (c) 200 W/m² → kcal/(h·m²)
val_c = 200 * Wm2_para_kcalhm2
print(f"(c) 200 W/m² = {val_c:.1f} kcal/(h·m²)")

print(f"\n📋 Fatores de conversão úteis:")
print(f"  • 1 kcal = 4,184 kJ")
print(f"  • 1 kW ≈ 3413 BTU/h")
print(f"  • 1 W/m² ≈ 0,86 kcal/(h·m²)")
print(f"  • 1 cv ≈ 735,5 W ≈ 0,986 hp")
```

## 🧱 Seção 3: Resistência Térmica e Paredes Compostas

### Exercício 3.1: Parede Composta em Série

Calcule a resistência térmica total de uma parede composta por:
- Tijolo: $k = 0,7$ W/(m·K), $L = 15$ cm
- Poliuretano: $k = 0,025$ W/(m·K), $L = 5$ cm
- Gesso: $k = 0,5$ W/(m·K), $L = 2$ cm

Área $A = 10$ m².

```python
# ============================================================
# EXERCÍCIO 3.1: Parede Composta
# ============================================================

print("=" * 60)
print("EXERCÍCIO 3.1: PAREDE COMPOSTA EM SÉRIE")
print("=" * 60)

# Dados
camadas = {
    'Tijolo': {'L': 0.15, 'k': 0.7},
    'Poliuretano': {'L': 0.05, 'k': 0.025},
    'Gesso': {'L': 0.02, 'k': 0.5}
}
A = 10.0  # m²

print(f"\n📊 Camadas da parede:")
print(f"{'Camada':<15} {'L [cm]':<10} {'k [W/(m·K)]':<15}")
print("-" * 40)
for nome, dados in camadas.items():
    print(f"{nome:<15} {dados['L']*100:<10.1f} {dados['k']:<15}")

# Cálculo das resistências
R_total = 0
R_list = []

print(f"\n🧮 Resistências térmicas:")
print(f"{'Camada':<15} {'R_th [K/W]':<15} {'% do total':<10}")
print("-" * 40)

for nome, dados in camadas.items():
    R = dados['L'] / (dados['k'] * A)
    R_list.append((nome, R))
    R_total += R

for nome, R in R_list:
    percentual = (R / R_total) * 100
    print(f"{nome:<15} {R:<15.4f} {percentual:<10.1f}")

print("-" * 40)
print(f"{'TOTAL':<15} {R_total:<15.4f} 100.0")

print(f"\n💡 Interpretação:")
print(f"  A camada de poliuretano domina a resistência total")
print(f"  ({R_list[1][1]/R_total*100:.1f}%), mesmo sendo a segunda")
print(f"  mais fina. Isso porque sua condutividade é muito baixa.")
```

### Exercício 3.2: Perda Térmica em Tubo Isolado

Um tubo de aço ($k = 45$ W/(m·K)) com $D_i = 50$ mm e $D_e = 60$ mm transporta vapor a 150°C. O ar externo está a 25°C com $h_o = 15$ W/(m²·K). Calcule a perda térmica por metro de comprimento.

```python
# ============================================================
# EXERCÍCIO 3.2: Tubo Isolado
# ============================================================

print("=" * 60)
print("EXERCÍCIO 3.2: PERDA TÉRMICA EM TUBO ISOLADO")
print("=" * 60)

# Dados
k_aco = 45.0       # W/(m·K)
D_i = 0.050        # m
D_e = 0.060        # m
T_vapor = 150.0    # °C
T_ar = 25.0        # °C
h_o = 15.0         # W/(m²·K)

# Resistências por metro de comprimento
R_cond = np.log(D_e / D_i) / (2 * np.pi * k_aco)
R_conv = 1 / (h_o * np.pi * D_e)
R_total = R_cond + R_conv

print(f"\n📊 Resistências térmicas (por metro):")
print(f"  • Condução no aço: R_cond = {R_cond:.5f} K·m/W")
print(f"  • Convecção externa: R_conv = {R_conv:.4f} K·m/W")
print(f"  • Total: R_total = {R_total:.4f} K·m/W")

# Perda térmica
Q_linha = (T_vapor - T_ar) / R_total

print(f"\n🧮 Perda térmica por metro:")
print(f"  • Q/L = (T_vapor - T_ar) / R_total")
print(f"  • Q/L = ({T_vapor} - {T_ar}) / {R_total:.4f}")
print(f"  • Q/L = {Q_linha:.1f} W/m")

print(f"\n💡 Interpretação:")
print(f"  A convecção externa domina a resistência ({R_conv/R_total*100:.1f}%),")
print(f"  pois o aço tem alta condutividade e a parede é fina.")
print(f"  Para reduzir a perda, deve-se adicionar isolante térmico.")
```

## ☀️ Seção 4: Radiação Térmica

### Exercício 4.1: Radiação de Concreto para o Céu Noturno

Uma superfície de concreto ($\varepsilon = 0,90$) a 40°C troca calor por radiação com o céu noturno a -10°C. Calcule o fluxo radiativo líquido.

```python
# ============================================================
# EXERCÍCIO 4.1: Radiação Térmica
# ============================================================

print("=" * 60)
print("EXERCÍCIO 4.1: RADIAÇÃO PARA O CÉU NOTURNO")
print("=" * 60)

# Constantes
sigma = 5.67e-8    # W/(m²·K⁴) - Stefan-Boltzmann
eps = 0.90         # emissividade do concreto

# Temperaturas em Kelvin
T_s = 40 + 273.15      # K
T_ceu = -10 + 273.15   # K

# Fluxo radiativo
q_rad = eps * sigma * (T_s**4 - T_ceu**4)

# Coeficiente radiativo linearizado
h_rad = q_rad / (T_s - T_ceu)

print(f"\n📊 Dados do problema:")
print(f"  • Emissividade: ε = {eps}")
print(f"  • Temperatura da superfície: T_s = {T_s:.2f} K ({T_s-273.15:.0f}°C)")
print(f"  • Temperatura do céu: T_céu = {T_ceu:.2f} K ({T_ceu-273.15:.0f}°C)")
print(f"  • σ = {sigma:.2e} W/(m²·K⁴)")

print(f"\n🧮 Cálculo:")
print(f"  • q'' = ε·σ·(T_s⁴ - T_céu⁴)")
print(f"  • q'' = {eps} × {sigma:.2e} × ({T_s:.2f}⁴ - {T_ceu:.2f}⁴)")
print(f"  • q'' = {q_rad:.1f} W/m²")
print(f"\n  • h_rad (linearizado) = {h_rad:.2f} W/(m²·K)")

print(f"\n💡 Interpretação:")
print(f"  O fluxo radiativo ({q_rad:.0f} W/m²) é significativo,")
print(f"  comparável ao fluxo convectivo para h ≈ 10 W/(m²·K).")
print(f"  Isso mostra que a radiação não pode ser desprezada")
print(f"  em problemas ao ar livre, mesmo em temperaturas moderadas.")
```

## 📊 Seção 5: Visualização Comparativa

Vamos visualizar a dependência do fluxo radiativo com a temperatura.

```python
# ============================================================
# VISUALIZAÇÃO: RADIAÇÃO vs CONVECÇÃO
# ============================================================

# Temperatura da superfície variável
T_s_range = np.linspace(0, 200, 100) + 273.15  # K
T_inf = 25 + 273.15  # K (ambiente)

# Fluxo radiativo
q_rad_range = eps * sigma * (T_s_range**4 - T_inf**4)

# Fluxo convectivo (para comparação)
h_conv = 10.0  # W/(m²·K)
q_conv_range = h_conv * (T_s_range - T_inf)

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(T_s_range - 273.15, q_rad_range, 'r-', linewidth=2, label='Radiação (ε = 0,9)')
ax.plot(T_s_range - 273.15, q_conv_range, 'b-', linewidth=2, label=f'Convecção (h = {h_conv} W/(m²·K))')

ax.set_xlabel('Temperatura da Superfície [°C]', fontsize=12)
ax.set_ylabel('Fluxo de Calor [W/m²]', fontsize=12)
ax.set_title('Radiação vs. Convecção', fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
ax.axhline(y=0, color='gray', linestyle='--', alpha=0.5)
ax.axvline(x=25, color='gray', linestyle=':', alpha=0.5, label='T_ambiente')

plt.tight_layout()
plt.show()

print("\n💡 Observação:")
print("  • A radiação cresce com T⁴ (não-linear)")
print("  • A convecção cresce linearmente com ΔT")
print("  • Para T altas (> 100°C), a radiação DOMINA")
print("  • Para T próximas à ambiente, convecção pode dominar")
```
