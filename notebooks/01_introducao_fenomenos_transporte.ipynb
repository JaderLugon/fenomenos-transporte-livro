## 📚 A Analogia Fundamental

Os três fenômenos de transporte — **momentum**, **calor** e **massa** — são governados por leis constitutivas que compartilham a mesma estrutura matemática:

$$
\text{Fluxo} = -(\text{Difusividade}) \times (\text{Gradiente do Potencial})
$$

| Fenômeno | Fluxo | Potencial | Difusividade | Lei |
|----------|-------|-----------|--------------|-----|
| Momentum | $\tau$ [Pa] | $du/dy$ [s⁻¹] | $\mu$ [Pa·s] | Newton |
| Calor | $q''$ [W/m²] | $dT/dy$ [K/m] | $k$ [W/(m·K)] | Fourier |
| Massa | $J$ [kg/(m²·s)] | $dC/dy$ [kg/m⁴] | $D$ [m²/s] | Fick |

Essa unificação permite **transferir soluções** entre os fenômenos: se você entende condução térmica, entende difusão de massa!

## 🔬 Seção 1: Verificação Dimensional das Três Leis

A consistência dimensional é o primeiro teste de validade de qualquer equação física. Vamos verificar que as três leis constitutivas são dimensionalmente corretas.

### Exercício 1.1: Lei de Newton da Viscosidade

$$\tau = \mu \frac{du}{dy}$$

**Tarefa:** Verifique que as unidades do lado esquerdo [Pa] são iguais às do lado direito.

```
# ============================================================
# EXERCÍCIO 1.1: Verificação dimensional - Newton
# ============================================================

print("=" * 60)
print("VERIFICAÇÃO DIMENSIONAL - LEI DE NEWTON")
print("=" * 60)

# Unidades no SI
unidades = {
    "τ (tensão de cisalhamento)": "Pa = kg/(m·s²)",
    "μ (viscosidade dinâmica)": "Pa·s = kg/(m·s)",
    "du/dy (gradiente de velocidade)": "s⁻¹"
}

print("\n📋 Unidades de cada termo:")
for termo, unidade in unidades.items():
    print(f"  • {termo}: {unidade}")

print("\n🔍 Verificação:")
print("  [μ] × [du/dy] = [kg/(m·s)] × [s⁻¹]")
print("                = kg/(m·s²)")
print("                = [Pa] ✓")

print("\n✅ A Lei de Newton é DIMENSIONALMENTE CONSISTENTE!")
```

### Exercício 1.2: Lei de Fourier

$$q'' = -k \frac{dT}{dy}$$

**Tarefa:** Verifique dimensionalmente e calcule o fluxo de calor para um exemplo prático.

```
# ============================================================
# EXERCÍCIO 1.2: Lei de Fourier - Verificação e Cálculo
# ============================================================

print("=" * 60)
print("LEI DE FOURIER - VERIFICAÇÃO DIMENSIONAL")
print("=" * 60)

print("\n🔍 Verificação dimensional:")
print("  [k] × [dT/dy] = [W/(m·K)] × [K/m]")
print("                = W/m²")
print("                = [q''] ✓")

# Exemplo prático: parede de tijolo
print("\n" + "=" * 60)
print("EXEMPLO PRÁTICO: Parede de Tijolo")
print("=" * 60)

# Dados
k_tijolo = 0.7  # W/(m·K)
T_interna = 25  # °C
T_externa = 10  # °C
L = 0.15  # m (espessura)

# Cálculo
dT_dy = (T_externa - T_interna) / L  # K/m
q_fluxo = -k_tijolo * dT_dy  # W/m²

print(f"\n📊 Dados do problema:")
print(f"  • Condutividade do tijolo: k = {k_tijolo} W/(m·K)")
print(f"  • Temperatura interna: T₁ = {T_interna}°C")
print(f"  • Temperatura externa: T₂ = {T_externa}°C")
print(f"  • Espessura: L = {L} m")

print(f"\n🧮 Cálculo:")
print(f"  • Gradiente: dT/dy = ({T_externa} - {T_interna}) / {L} = {dT_dy:.2f} K/m")
print(f"  • Fluxo: q'' = -{k_tijolo} × ({dT_dy:.2f}) = {q_fluxo:.2f} W/m²")

print(f"\n💡 Interpretação:")
print(f"  O fluxo de calor é {abs(q_fluxo):.1f} W/m² saindo da parede")
print(f"  (sentido positivo indica fluxo no sentido +y, do quente para o frio)")
```

### Exercício 1.3: Lei de Fick

$$J = -D \frac{dC}{dy}$$

**Tarefa:** Verifique dimensionalmente e calcule o fluxo de um poluente em um rio.

```
# ============================================================
# EXERCÍCIO 1.3: Lei de Fick - Verificação e Cálculo
# ============================================================

print("=" * 60)
print("LEI DE FICK - VERIFICAÇÃO DIMENSIONAL")
print("=" * 60)

print("\n🔍 Verificação dimensional:")
print("  [D] × [dC/dy] = [m²/s] × [kg/m⁴]")
print("                = kg/(m²·s)")
print("                = [J] ✓")

# Exemplo prático: difusão de poluente
print("\n" + "=" * 60)
print("EXEMPLO PRÁTICO: Difusão de Poluente")
print("=" * 60)

# Dados
D_pol = 1.5e-9  # m²/s (difusividade molecular típica)
C_margem = 0.050  # kg/m³ (50 mg/L na margem)
C_centro = 0.005  # kg/m³ (5 mg/L no centro)
distancia = 10  # m

# Cálculo
dC_dy = (C_centro - C_margem) / distancia  # kg/m⁴
J_fluxo = -D_pol * dC_dy  # kg/(m²·s)

print(f"\n📊 Dados do problema:")
print(f"  • Difusividade molecular: D = {D_pol:.2e} m²/s")
print(f"  • Concentração na margem: C₁ = {C_margem*1000:.1f} mg/L")
print(f"  • Concentração no centro: C₂ = {C_centro*1000:.1f} mg/L")
print(f"  • Distância: Δy = {distancia} m")

print(f"\n🧮 Cálculo:")
print(f"  • Gradiente: dC/dy = {dC_dy:.2e} kg/m⁴")
print(f"  • Fluxo: J = {J_fluxo:.2e} kg/(m²·s)")

print(f"\n💡 Interpretação:")
print(f"  A difusão molecular é MUITO lenta!")
print(f"  Em rios, a dispersão hidrodinâmica (E ~ 1-100 m²/s) domina")
print(f"  sobre a difusão molecular (D ~ 10⁻⁹ m²/s) por ~10 ordens de grandeza!")
```

## 📐 Seção 2: Propriedades Extensivas e Intensivas

Uma propriedade é:
- **Extensiva ($B$)**: depende da quantidade de matéria (ex: massa, volume, energia)
- **Intensiva ($\beta$)**: independente da quantidade (ex: temperatura, pressão, densidade)

A relação entre ambas é:
$$\beta = \frac{dB}{dm} \quad \text{ou} \quad B = \int_V \rho \beta \, dV$$

## 🔢 Seção 3: Números Adimensionais Fundamentais

Os números adimensionais quantificam a competição entre mecanismos físicos. Eles surgem naturalmente da adimensionalização das equações governantes.

### Tabela de Números Adimensionais

| Número | Fórmula | Significado Físico |
|--------|---------|-------------------|
| Reynolds (Re) | $UL/\nu$ | Inércia/Viscosidade |
| Péclet (Pe) | $UL/D$ | Advecção/Difusão |
| Prandtl (Pr) | $\nu/\alpha$ | Camada hidrodinâmica/térmica |
| Schmidt (Sc) | $\nu/D$ | Camada hidrodinâmica/mássica |
| Froude (Fr) | $U/\sqrt{gL}$ | Inércia/Gravidade |

### Exercício 3.1: Calculadora de Números Adimensionais

```
# ============================================================
# EXERCÍCIO 3.1: Calculadora de Números Adimensionais
# ============================================================

def calcular_reynolds(U, L, nu):
    """Calcula o número de Reynolds"""
    return U * L / nu

def calcular_peclet(U, L, D):
    """Calcula o número de Péclet"""
    return U * L / D

def calcular_prandtl(nu, alpha):
    """Calcula o número de Prandtl"""
    return nu / alpha

def calcular_schmidt(nu, D):
    """Calcula o número de Schmidt"""
    return nu / D

def classificar_regime(Re, Re_lam=2000, Re_turb=2400):
    """Classifica o regime de escoamento"""
    if Re < Re_lam:
        return "LAMINAR"
    elif Re > Re_turb:
        return "TURBULENTO"
    else:
        return "TRANSIÇÃO"

def classificar_transporte(Pe):
    """Classifica o regime de transporte"""
    if Pe > 10:
        return "ADVECÇÃO DOMINANTE"
    elif Pe < 0.1:
        return "DIFUSÃO DOMINANTE"
    else:
        return "MISTO (advecção ≈ difusão)"

# Exemplo 1: Água em tubo residencial
print("=" * 60)
print("EXEMPLO 1: Água em Tubo Residencial")
print("=" * 60)

U = 1.0  # m/s
D_tubo = 0.025  # m (25 mm)
nu_agua = 1.0e-6  # m²/s
alpha_agua = 1.4e-7  # m²/s
D_sal = 1.5e-9  # m²/s (difusividade do sal em água)

Re = calcular_reynolds(U, D_tubo, nu_agua)
Pe_termico = calcular_peclet(U, D_tubo, alpha_agua)
Pe_massico = calcular_peclet(U, D_tubo, D_sal)
Pr = calcular_prandtl(nu_agua, alpha_agua)
Sc = calcular_schmidt(nu_agua, D_sal)

print(f"\n📊 Parâmetros:")
print(f"  • Velocidade: U = {U} m/s")
print(f"  • Diâmetro: D = {D_tubo*1000} mm")
print(f"  • ν (água) = {nu_agua:.2e} m²/s")

print(f"\n🔢 Números Adimensionais:")
print(f"  • Re = {Re:.0f} → {classificar_regime(Re)}")
print(f"  • Pe_térmico = {Pe_termico:.0f} → {classificar_transporte(Pe_termico)}")
print(f"  • Pe_mássico = {Pe_massico:.0f} → {classificar_transporte(Pe_massico)}")
print(f"  • Pr = {Pr:.2f}")
print(f"  • Sc = {Sc:.0f}")

print(f"\n💡 Interpretação:")
print(f"  • Pr > 1 → camada limite térmica mais fina que a hidrodinâmica")
print(f"  • Sc >> 1 → sal difunde MUITO mais devagar que o momentum")
print(f"  • Por isso corantes demoram tanto para se misturar em água parada!")

# Exemplo 2: Rio Macaé
print("\n" + "=" * 60)
print("EXEMPLO 2: Rio Macaé (RJ)")
print("=" * 60)

U_rio = 0.20  # m/s
L_rio = 42.2  # m (largura)
E_L = 12.57  # m²/s (dispersão longitudinal)
E_T = 0.00431  # m²/s (dispersão transversal)

Pe_L = calcular_peclet(U_rio, L_rio, E_L)
Pe_T = calcular_peclet(U_rio, L_rio, E_T)

print(f"\n📊 Parâmetros:")
print(f"  • Velocidade: U = {U_rio} m/s")
print(f"  • Largura: B = {L_rio} m")
print(f"  • E_L = {E_L} m²/s | E_T = {E_T} m²/s")

print(f"\n🔢 Números de Péclet:")
print(f"  • Pe_L (longitudinal) = {Pe_L:.2f} → {classificar_transporte(Pe_L)}")
print(f"  • Pe_T (transversal) = {Pe_T:.0f} → {classificar_transporte(Pe_T)}")

print(f"\n💡 Interpretação:")
print(f"  • Pe_L ≈ 1 → advecção e dispersão longitudinal são comparáveis")
print(f"  • Pe_T >> 1 → dispersão transversal é muito lenta")
print(f"  • Por isso plumas de poluentes se alongam antes de se misturar!")
```

## ⚡ Seção 4: Analogia com Circuitos Elétricos

Problemas de condução térmica podem ser modelados como circuitos elétricos:

| Circuito Elétrico | Circuito Térmico |
|-------------------|------------------|
| Diferença de potencial ΔV [V] | Diferença de temperatura ΔT [K] |
| Corrente I [A] | Fluxo de calor Q [W] |
| Resistência R [Ω] | Resistência térmica R_th [K/W] |

**Lei de Ohm:** $I = \Delta V / R$  
**Lei de Fourier:** $\dot{Q} = \Delta T / R_{th}$

### Exercício 4.1: Parede Composta como Circuito em Série

```
# ============================================================
# EXERCÍCIO 4.1: Circuito Térmico em Série
# ============================================================

print("=" * 60)
print("ANALOGIA ELÉTRICA - PAREDE COMPOSTA")
print("=" * 60)

# Dados da parede (forno de padaria - caso do Cap. 7)
camadas = {
    "Aço interno": {"L": 0.003, "k": 45},
    "Lã de vidro": {"L": 0.0108, "k": 0.04},
    "Aço externo": {"L": 0.002, "k": 45}
}

A = 2.0  # m² (área)
T_interna = 250  # °C
T_externa = 25  # °C (ambiente)

print("\n📊 Camadas da parede:")
print(f"{'Camada':<15} {'L [mm]':<10} {'k [W/(m·K)]':<15}")
print("-" * 40)
for nome, dados in camadas.items():
    print(f"{nome:<15} {dados['L']*1000:<10.1f} {dados['k']:<15}")

# Cálculo das resistências
R_total = 0
print("\n🔍 Cálculo das resistências térmicas:")
print(f"{'Camada':<15} {'R_th [K/W]':<15} {'% do total':<10}")
print("-" * 40)

for nome, dados in camadas.items():
    R = dados['L'] / (dados['k'] * A)
    R_total += R
    print(f"{nome:<15} {R:<15.4f} {'(calculando...)':<10}")

# Recalcular para mostrar percentuais
print("\n📊 Contribuição percentual:")
for nome, dados in camadas.items():
    R = dados['L'] / (dados['k'] * A)
    percentual = (R / R_total) * 100
    print(f"  • {nome}: {percentual:.1f}%")

# Cálculo do fluxo
Q = (T_interna - T_externa) / R_total

print(f"\n🧮 Resultados:")
print(f"  • R_total = {R_total:.4f} K/W")
print(f"  • ΔT = {T_interna - T_externa} K")
print(f"  • Fluxo de calor Q = {Q:.1f} W")
print(f"  • Densidade de fluxo q'' = {Q/A:.1f} W/m²")

print(f"\n💡 Insight importante:")
print(f"  A camada de lã de vidro (isolante) domina a resistência,")
print(f"  mesmo sendo a mais FINA em espessura absoluta!")
print(f"  Isso porque k_lã << k_aço (0.04 vs 45 W/(m·K))")
```

## 🔄 Seção 5: Conversão de Unidades

A consistência dimensional é crítica em engenharia. Vamos praticar conversões.

### Exercício 5.1: Conversor de Unidades Térmicas

```
# ============================================================
# EXERCÍCIO 5.1: Conversor de Unidades
# ============================================================

def converter_energia(valor, de_unidade, para_unidade):
    """Converte entre unidades de energia"""
    # Fatores de conversão para Joules
    fatores = {
        "J": 1,
        "kJ": 1e3,
        "MJ": 1e6,
        "cal": 4.184,
        "kcal": 4184,
        "BTU": 1055,
        "kWh": 3.6e6
    }
    
    valor_J = valor * fatores[de_unidade]
    resultado = valor_J / fatores[para_unidade]
    
    return resultado

def converter_potencia(valor, de_unidade, para_unidade):
    """Converte entre unidades de potência"""
    fatores = {
        "W": 1,
        "kW": 1e3,
        "MW": 1e6,
        "cv": 735.5,
        "hp": 745.7,
        "BTU/h": 0.293
    }
    
    valor_W = valor * fatores[de_unidade]
    resultado = valor_W / fatores[para_unidade]
    
    return resultado

# Exemplos de conversão
print("=" * 60)
print("CONVERSOR DE UNIDADES TÉRMICAS")
print("=" * 60)

print("\n📊 Conversões de ENERGIA:")
print(f"  • 500 kcal = {converter_energia(500, 'kcal', 'kJ'):.1f} kJ")
print(f"  • 1 kWh = {converter_energia(1, 'kWh', 'MJ'):.2f} MJ")
print(f"  • 1 BTU = {converter_energia(1, 'BTU', 'J'):.1f} J")

print("\n📊 Conversões de POTÊNCIA:")
print(f"  • 3 kW = {converter_potencia(3, 'kW', 'BTU/h'):.1f} BTU/h")
print(f"  • 100 cv = {converter_potencia(100, 'cv', 'kW'):.1f} kW")
print(f"  • 1 hp = {converter_potencia(1, 'hp', 'W'):.1f} W")

print("\n📊 Conversões de FLUXO TÉRMICO:")
# 1 kcal/(h·m²) = 1.163 W/m²
print(f"  • 200 W/m² = {200/1.163:.1f} kcal/(h·m²)")
print(f"  • 1 BTU/(h·ft²) = {3.154:.3f} W/m²")
```
