"""))

```python
# [CELL 4] 🎓 Exercício Graduação: Cálculo de Difusividades
display(Markdown(r"""
### 🎓 Exercício 1: Cálculo de Difusividades Características

**Enunciado:** Para água a 25°C, calcule:
a) Difusividade cinemática $\nu$ (transporte de momentum)
b) Difusividade térmica $\alpha$ (transporte de calor)
c) Razão $\nu/\alpha$ (número de Prandtl)

**Dados:**
- $\mu = 8.90 \times 10^{-4}$ Pa·s
- $\rho = 997$ kg/m³
- $k = 0.607$ W/(m·K)
- $c_p = 4182$ J/(kg·K)
"""))

# Solução passo a passo
mu = 8.90e-4      # Pa·s
rho = 997          # kg/m³
k = 0.607          # W/(m·K)
cp = 4182          # J/(kg·K)

# a) Difusividade cinemática
nu = mu / rho
print(f"a) ν = μ/ρ = {mu:.3e} / {rho} = {nu:.3e} m²/s")

# b) Difusividade térmica
alpha = k / (rho * cp)
print(f"b) α = k/(ρ·cp) = {k} / ({rho}×{cp}) = {alpha:.3e} m²/s")

# c) Número de Prandtl
Pr = nu / alpha
print(f"c) Pr = ν/α = {nu:.3e} / {alpha:.3e} = {Pr:.2f}")

# Visualização comparativa
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

# Gráfico de barras das difusividades
diffusivities = [nu, alpha, 2.1e-9]  # D para O₂ em água
labels = ['ν (momentum)', 'α (calor)', 'D (massa)']
colors = ['#2E86AB', '#A23B72', '#F18F01']

ax[0].bar(labels, diffusivities, color=colors, edgecolor='black')
ax[0].set_ylabel('Difusividade [m²/s]')
ax[0].set_yscale('log')
ax[0].set_title('Comparação de Difusividades (água, 25°C)')
ax[0].grid(True, alpha=0.3)

# Gráfico do número de Prandtl
Pr_values = np.linspace(0.1, 10, 100)
ax[1].axvline(x=Pr, color='red', linestyle='--', label=f'Pr = {Pr:.2f} (água)')
ax[1].fill_betweenx([0, 1], 0.1, 1, alpha=0.2, color='blue', label='Pr < 1 (metais)')
ax[1].fill_betweenx([0, 1], 1, 10, alpha=0.2, color='orange', label='Pr > 1 (óleos)')
ax[1].set_xlabel('Número de Prandtl')
ax[1].set_ylabel('Regime')
ax[1].set_title('Interpretação do Número de Prandtl')
ax[1].legend()
ax[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

display(Markdown(f"""
> **Interpretação:** Para água, Pr ≈ {Pr:.1f} > 1, indicando que a camada limite hidrodinâmica 
> é mais espessa que a camada limite térmica. Isso significa que o momentum se difunde 
> mais rapidamente que o calor no fluido.
"""))
```

```python
# [CELL 5] 🔬 Aprofundamento: Forma Geral da Equação de Transporte
display(Markdown(r"""
## 🔬 Derivação da Equação Geral de Transporte

### Teorema do Transporte de Reynolds (TTR)

Para uma propriedade extensiva $B$ com densidade intensiva $\beta = dB/dm$:

$$\frac{DB}{Dt} = \frac{\partial}{\partial t}\int_{VC} \rho\beta \, dV + \oint_{SC} \rho\beta (\mathbf{u}\cdot\mathbf{n}) \, dA$$

### Forma Diferencial de Conservação:

$$\frac{\partial(\rho\beta)}{\partial t} + \nabla\cdot(\rho\beta\mathbf{u}) = \nabla\cdot(\Gamma_\beta\nabla\beta) + \dot{g}_\beta$$

### Casos Particulares:

| Propriedade | $\beta$ | $\Gamma_\beta$ | $\dot{g}_\beta$ | Equação Resultante |
|-------------|---------|----------------|-----------------|-------------------|
| Massa total | 1 | — | 0 | $\frac{\partial\rho}{\partial t} + \nabla\cdot(\rho\mathbf{u}) = 0$ |
| Momentum | $\mathbf{u}$ | $\mu$ | $\rho\mathbf{g}$ | Navier-Stokes |
| Energia | $c_p T$ | $k$ | $\dot{q}_{gen}$ | $\rho c_p \frac{DT}{Dt} = \nabla\cdot(k\nabla T) + \dot{q}_{gen}$ |
| Espécie $i$ | $C_i$ | $\rho D_i$ | $\dot{r}_i$ | $\frac{\partial C_i}{\partial t} + \nabla\cdot(C_i\mathbf{u}) = \nabla\cdot(\rho D_i\nabla C_i) + \dot{r}_i$ |
"""))

# Função genérica para calcular termos da equação de transporte
def transporte_geral(beta, Gamma, u, grad_beta, source_term=0, rho=1.0):
    """
    Calcula o termo de transporte difusivo para a forma geral.
    
    Parâmetros:
    -----------
    beta : float ou array
        Propriedade intensiva transportada
    Gamma : float
        Coeficiente de difusão/condutividade
    u : float ou array
        Campo de velocidade
    grad_beta : float ou array
        Gradiente da propriedade
    source_term : float, opcional
        Termo fonte/sumidouro
    rho : float, opcional
        Densidade do meio
    
    Retorna:
    --------
    flux_diff : float ou array
        Fluxo difusivo = -Gamma * grad_beta
    flux_adv : float ou array
        Fluxo advectivo = rho * beta * u
    """
    flux_diff = -Gamma * grad_beta
    flux_adv = rho * beta * u
    return flux_diff, flux_adv

# Exemplo de uso: comparação entre os três fenômenos
print("=== Exemplo: Fluxos em parede plana (1D) ===\n")

# Condições comuns
grad = -100  # Gradiente genérico [unidade/m]
u = 0.1      # Velocidade [m/s]
rho = 1000   # Densidade [kg/m³]

# Momentum
mu = 1e-3
tau_diff, tau_adv = transporte_geral(beta=u, Gamma=mu, u=u, grad_beta=grad, rho=rho)
print(f"Momentum (Newton):")
print(f"  Fluxo difusivo (tensão): τ_diff = {tau_diff:.3e} Pa")
print(f"  Fluxo advectivo: τ_adv = {tau_adv:.3e} Pa·m/s\n")

# Calor
k = 0.6
cp = 4180
T = 300  # Temperatura de referência
q_diff, q_adv = transporte_geral(beta=cp*T, Gamma=k, u=u, grad_beta=grad, rho=rho)
print(f"Calor (Fourier):")
print(f"  Fluxo difusivo: q''_diff = {q_diff:.3e} W/m²")
print(f"  Fluxo advectivo: q''_adv = {q_adv:.3e} W/m²\n")

# Massa
D = 2e-9
C = 0.01  # Concentração [kg/m³]
J_diff, J_adv = transporte_geral(beta=C, Gamma=rho*D, u=u, grad_beta=grad, rho=rho)
print(f"Massa (Fick):")
print(f"  Fluxo difusivo: J_diff = {J_diff:.3e} kg/(m²·s)")
print(f"  Fluxo advectivo: J_adv = {J_adv:.3e} kg/(m²·s)")
```

```python
# [CELL 6] 🎓🔬 Números Adimensionais e Diagrama de Regimes
display(Markdown(r"""
## 📊 Números Adimensionais Fundamentais

Os fenômenos de transporte são governados pela **competição entre mecanismos**, quantificada por números adimensionais:

| Número | Fórmula | Interpretação Física |
|--------|---------|---------------------|
| **Reynolds (Re)** | $Re = \frac{UL}{\nu}$ | Inércia / Viscosidade → Diagnóstico de turbulência |
| **Péclet (Pe)** | $Pe = \frac{UL}{\Gamma}$ | Advecção / Difusão → Mecanismo dominante |
| **Prandtl (Pr)** | $Pr = \frac{\nu}{\alpha}$ | Camada limite hidrodinâmica / térmica |
| **Schmidt (Sc)** | $Sc = \frac{\nu}{D}$ | Camada limite hidrodinâmica / mássica |
| **Nusselt (Nu)** | $Nu = \frac{hL}{k_f}$ | Convecção / Condução no fluido |

### Diagrama de Regimes Re × Pe:

O diagrama abaixo classifica escoamentos combinando:
- **Eixo X**: Número de Reynolds (laminar → turbulento)
- **Eixo Y**: Número de Péclet (difusão → advecção dominante)
"""))

# Criar diagrama de regimes Re x Pe
def criar_diagrama_regimes():
    """Cria o diagrama de regimes baseado em números adimensionais."""
    
    # Criar grade de valores
    Re = np.logspace(0, 6, 200)  # 1 a 1e6
    Pe = np.logspace(-2, 4, 200)  # 0.01 a 10000
    
    # Criar meshgrid
    Re_grid, Pe_grid = np.meshgrid(Re, Pe)
    
    # Definir regiões
    # Limites de transição laminar-turbulento (aproximados)
    Re_lam = 2000
    Re_turb = 4000
    
    # Criar máscara para coloração
    region = np.zeros_like(Re_grid)
    
    # Região 1: Difusão + Laminar (Pe < 1, Re < 2000)
    mask1 = (Pe_grid < 1) & (Re_grid < Re_lam)
    region[mask1] = 1
    
    # Região 2: Difusão + Transição (Pe < 1, 2000 < Re < 4000)
    mask2 = (Pe_grid < 1) & (Re_grid >= Re_lam) & (Re_grid < Re_turb)
    region[mask2] = 2
    
    # Região 3: Difusão + Turbulento (Pe < 1, Re > 4000)
    mask3 = (Pe_grid < 1) & (Re_grid >= Re_turb)
    region[mask3] = 3
    
    # Região 4: Advecção + Laminar (Pe > 1, Re < 2000)
    mask4 = (Pe_grid >= 1) & (Re_grid < Re_lam)
    region[mask4] = 4
    
    # Região 5: Advecção + Transição (Pe > 1, 2000 < Re < 4000)
    mask5 = (Pe_grid >= 1) & (Re_grid >= Re_lam) & (Re_grid < Re_turb)
    region[mask5] = 5
    
    # Região 6: Advecção + Turbulento (Pe > 1, Re > 4000)
    mask6 = (Pe_grid >= 1) & (Re_grid >= Re_turb)
    region[mask6] = 6
    
    # Plotar
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Cores para as regiões
    colors = ['#E3F2FD', '#BBDEFB', '#90CAF9',  # Difusão (azuis)
              '#FFF3E0', '#FFE0B2', '#FFCC80']   # Advecção (laranjas)
    
    labels_region = [
        'Difusão + Laminar', 'Difusão + Transição', 'Difusão + Turbulento',
        'Advecção + Laminar', 'Advecção + Transição', 'Advecção + Turbulento'
    ]
    
    for i in range(1, 7):
        ax.contourf(Re_grid, Pe_grid, region, levels=[i-0.5, i+0.5], 
                   colors=[colors[i-1]], alpha=0.3)
    
    # Linhas de referência
    ax.axvline(x=Re_lam, color='gray', linestyle='--', alpha=0.5, label='Re = 2000')
    ax.axvline(x=Re_turb, color='gray', linestyle='--', alpha=0.5, label='Re = 4000')
    ax.axhline(y=1, color='black', linestyle='-', linewidth=1.5, label='Pe = 1')
    
    # Pontos de exemplo (Rio Macaé, Tubulação, Microcanal)
    exemplos = {
        'Rio Macaé': {'Re': 1.4e5, 'Pe': 3.5e4, 'color': 'green', 'marker': 'o'},
        'Tubulação': {'Re': 6.4e4, 'Pe': 8.5e4, 'color': 'blue', 'marker': 's'},
        'Microcanal': {'Re': 50, 'Pe': 0.5, 'color': 'red', 'marker': '^'}
    }
    
    for nome, dados in exemplos.items():
        ax.plot(dados['Re'], dados['Pe'], marker=dados['marker'], 
               color=dados['color'], markersize=10, label=nome, 
               markeredgecolor='black', markeredgewidth=1.5)
    
    # Configurações do gráfico
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel('Número de Reynolds (Re)', fontsize=12)
    ax.set_ylabel('Número de Péclet (Pe)', fontsize=12)
    ax.set_title('Diagrama de Regimes: Classificação de Escoamentos', fontsize=14, fontweight='bold')
    
    # Legenda
    ax.legend(loc='lower right', fontsize=9)
    ax.grid(True, alpha=0.3, which='both')
    
    # Texto explicativo
    ax.text(0.02, 0.98, 'Pe > 1: Advecção domina\nPe < 1: Difusão domina', 
           transform=ax.transAxes, fontsize=9, 
           bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.3),
           verticalalignment='top')
    
    plt.tight_layout()
    return fig, ax

fig, ax = criar_diagrama_regimes()
plt.show()

display(Markdown("""
> **Como usar na prática (🎓 Graduação):**
> 1. Calcule **Re** primeiro para saber se o escoamento é laminar ou turbulento
> 2. Calcule **Pe** para decidir se pode desprezar difusão molecular
> 3. Em rios, Pe varia de 10³ a 10⁶ → advecção domina o transporte de poluentes
> 4. Use **Pr** e **Sc** para comparar espessuras de camadas limite
"""))
```

```python
# [CELL 7] 🔬 Exercício Avançado: Adimensionalização da Equação de Advecção-Difusão
display(Markdown(r"""
### 🔬 Exercício Avançado: Adimensionalização e Limites Assintóticos

**Enunciado:** Adimensionalize a equação de advecção-difusão 1D:

$$\frac{\partial C}{\partial t} + U\frac{\partial C}{\partial x} = E\frac{\partial^2 C}{\partial x^2}$$

Mostre que o único parâmetro adimensional resultante é o número de Péclet.

**Solução Guiada:**
"""))

# Definir escalas características
L = 1000      # Comprimento característico [m]
U = 0.2       # Velocidade característica [m/s]
C0 = 0.01     # Concentração de referência [kg/m³]
E = 12.6      # Coeficiente de dispersão [m²/s]

# Calcular número de Péclet
Pe = U * L / E
print(f"Parâmetros de escala:")
print(f"  L = {L} m, U = {U} m/s, E = {E} m²/s")
print(f"  → Pe = UL/E = {Pe:.2f}")
print()

# Definir variáveis adimensionais
print("Variáveis adimensionais:")
print(f"  x* = x/L, t* = t·U/L, C* = C/C₀")
print()

# Mostrar transformação das derivadas
display(Markdown(r"""
**Transformação das derivadas:**

$$\frac{\partial}{\partial t} = \frac{U}{L}\frac{\partial}{\partial t^*}, \quad 
\frac{\partial}{\partial x} = \frac{1}{L}\frac{\partial}{\partial x^*}, \quad
\frac{\partial^2}{\partial x^2} = \frac{1}{L^2}\frac{\partial^2}{\partial x^{*2}}$$

**Substituindo na equação original:**

$$\frac{U}{L}C_0\frac{\partial C^*}{\partial t^*} + U\frac{C_0}{L}\frac{\partial C^*}{\partial x^*} = E\frac{C_0}{L^2}\frac{\partial^2 C^*}{\partial x^{*2}}$$

**Dividindo por $UC_0/L$:**

$$\frac{\partial C^*}{\partial t^*} + \frac{\partial C^*}{\partial x^*} = \frac{E}{UL}\frac{\partial^2 C^*}{\partial x^{*2}} = \frac{1}{Pe}\frac{\partial^2 C^*}{\partial x^{*2}}$$

✅ **Resultado:** A equação adimensionalizada depende apenas de **Pe**.
"""))

# Visualizar limites assintóticos
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Solução para Pe → 0 (difusão pura)
x = np.linspace(-5, 5, 200)
t_vals = [0.1, 0.5, 1.0, 2.0]

for t in t_vals:
    # Solução fundamental da equação do calor
    C_diff = np.exp(-x**2 / (4*t)) / np.sqrt(4*np.pi*t)
    axes[0].plot(x, C_diff, label=f't* = {t}', linewidth=2)

axes[0].set_xlabel('x*')
axes[0].set_ylabel('C*')
axes[0].set_title('Limite Pe → 0: Difusão Pura')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Solução para Pe → ∞ (advecção pura)
for t in t_vals:
    # Solução de onda viajante (aproximação)
    C_adv = np.exp(-(x - t)**2 / 0.1)  # Pulso que se move
    axes[1].plot(x, C_adv, label=f't* = {t}', linewidth=2)

axes[1].set_xlabel('x*')
axes[1].set_ylabel('C*')
axes[1].set_title('Limite Pe → ∞: Advecção Pura')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

display(Markdown(f"""
> **Interpretação dos limites (🔬 Pós-Graduação):**
> - **Pe → 0**: Equação parabólica (difusão) → esquemas centrados são estáveis
> - **Pe → ∞**: Equação hiperbólica (advecção) → requer esquemas upwind/TVD
> - **Pe ≈ {Pe:.0f}** (caso do Rio Macaé): Regime misto → cuidado com oscilações numéricas!
"""))
```

```python
# [CELL 8] 🎓 Aplicação Prática: Estimativa de Tempos de Transporte
display(Markdown(r"""
### 🎓 Aplicação Prática: Tempos Característicos de Transporte

**Problema:** Um poluente é lançado em um rio com as seguintes características:
- Velocidade média: $U = 0.2$ m/s
- Coeficiente de dispersão longitudinal: $E_L = 12.6$ m²/s
- Distância até ponto de monitoramento: $L = 5$ km

**Perguntas:**
1. Qual o tempo de viagem por advecção?
2. Qual a escala de tempo para dispersão longitudinal?
3. Qual mecanismo domina o transporte?
"""))

# Dados do problema
U = 0.2           # m/s
EL = 12.6         # m²/s
L = 5000          # m (5 km)

# 1. Tempo de advecção
t_adv = L / U
print(f"1. Tempo de viagem por advecção:")
print(f"   t_adv = L/U = {L} m / {U} m/s = {t_adv:.0f} s ≈ {t_adv/3600:.2f} h")
print()

# 2. Tempo de dispersão (escala difusiva)
t_diff = L**2 / (2 * EL)
print(f"2. Escala de tempo para dispersão:")
print(f"   t_diff = L²/(2·E) = {L}² / (2×{EL}) = {t_diff:.0f} s ≈ {t_diff/3600:.2f} h")
print()

# 3. Número de Péclet e dominância
Pe_local = U * L / EL
print(f"3. Número de Péclet local:")
print(f"   Pe = UL/E = {U}×{L}/{EL} = {Pe_local:.1f}")
print()

if Pe_local > 10:
    dominancia = "✅ ADVECÇÃO DOMINA (Pe >> 1)"
elif Pe_local < 0.1:
    dominancia = "✅ DIFUSÃO DOMINA (Pe << 1)"
else:
    dominancia = "⚠️ REGIME MISTO (Pe ≈ 1)"

print(f"   → {dominancia}")
print()

# Visualização: evolução temporal da pluma
t_hours = np.array([1, 2, 4, 8])  # horas
t_seconds = t_hours * 3600

fig, ax = plt.subplots(figsize=(10, 5))

for t in t_seconds:
    # Posição do pico por advecção
    x_pico = U * t
    
    # Largura da pluma por dispersão (desvio padrão)
    sigma = np.sqrt(2 * EL * t)
    
    # Perfil gaussiano normalizado
    x = np.linspace(max(0, x_pico - 4*sigma), x_pico + 4*sigma, 300)
    C = np.exp(-(x - x_pico)**2 / (2 * sigma**2))
    
    ax.plot(x/1000, C, label=f't = {t/3600:.0f} h', linewidth=2)
    ax.axvline(x=x_pico/1000, color='gray', linestyle=':', alpha=0.3)

ax.set_xlabel('Distância a jusante [km]')
ax.set_ylabel('Concentração normalizada C/C₀')
ax.set_title(f'Evolução da Pluma de Poluente (U={U} m/s, E_L={EL} m²/s)')
ax.legend()
ax.grid(True, alpha=0.3)
ax.axvline(x=L/1000, color='red', linestyle='--', label=f'Ponto de monitoramento ({L/1000} km)')

plt.tight_layout()
plt.show()

display(Markdown(f"""
> **Conclusão prática:**
> - O poluente leva **~{t_adv/3600:.1f} horas** para chegar ao ponto de monitoramento por advecção
> - A dispersão alarga a pluma com σ ≈ √(2·E·t) → após {t_hours[-1]} h: σ ≈ {np.sqrt(2*EL*t_seconds[-1])/1000:.2f} km
> - Como **Pe = {Pe_local:.0f} >> 1**, modelos simplificados podem desprezar difusão molecular para estimativas de tempo de chegada
> - ⚠️ Para previsão de concentrações máximas, a dispersão **não pode** ser desprezada!
"""))
```

```python
# [CELL 9] 🔬 Desafio Computacional: Função Genérica de Transporte
display(Markdown(r"""
### 🔬 Desafio Computacional: Implementação de Solver Genérico

**Objetivo:** Criar uma função Python que resolva a equação de transporte 1D para qualquer fenômeno (momentum, calor ou massa), usando o método das diferenças finitas explícito.

**Equação alvo:**
$$\frac{\partial \phi}{\partial t} + U\frac{\partial \phi}{\partial x} = \Gamma\frac{\partial^2 \phi}{\partial x^2} + S$$

onde $\phi$ pode ser: velocidade $u$, temperatura $T$, ou concentração $C$.
"""))

def solver_transporte_1d(phi0, U, Gamma, S, L, T_final, 
                         dx=1.0, dt=None, CFL=0.8, boundary='periodic'):
    """
    Solver 1D explícito para equação de advecção-difusão com fonte.
    
    Parâmetros:
    -----------
    phi0 : array
        Condição inicial da propriedade transportada
    U : float
        Velocidade de advecção [m/s]
    Gamma : float
        Coeficiente de difusão [m²/s]
    S : float ou função(x, t)
        Termo fonte/sumidouro
    L : float
        Comprimento do domínio [m]
    T_final : float
        Tempo final da simulação [s]
    dx : float, opcional
        Espaçamento da malha [m] (default: 1.0)
    dt : float, opcional
        Passo de tempo [s] (se None, calculado via CFL)
    CFL : float, opcional
        Número CFL para estabilidade (default: 0.8)
    boundary : str, opcional
        Tipo de condição de contorno: 'periodic', 'dirichlet', 'neumann'
    
    Retorna:
    --------
    x : array
        Vetor de posições
    t : array
        Vetor de tempos
    phi : array 2D
        Solução phi(x, t)
    """
    
    # Discretização espacial
    Nx = int(L / dx) + 1
    x = np.linspace(0, L, Nx)
    
    # Condição de estabilidade CFL + difusão
    if dt is None:
        dt_adv = CFL * dx / abs(U) if U != 0 else np.inf
        dt_diff = CFL * dx**2 / (2 * Gamma) if Gamma > 0 else np.inf
        dt = min(dt_adv, dt_diff, 1.0)  # Limitar dt máximo
    
    # Vetor de tempos
    Nt = int(T_final / dt) + 1
    t = np.linspace(0, T_final, Nt)
    
    # Inicializar solução
    phi = np.zeros((Nt, Nx))
    phi[0, :] = phi0(x) if callable(phi0) else np.full(Nx, phi0)
    
    # Loop temporal
    for n in range(Nt - 1):
        phi_n = phi[n, :].copy()
        
        # Termos de fonte
        source = S(x, t[n]) if callable(S) else np.full(Nx, S)
        
        # Diferenças finitas (esquema upwind para advecção + centrado para difusão)
        for i in range(1, Nx - 1):
            # Derivada espacial (upwind)
            if U >= 0:
                dphi_dx = (phi_n[i] - phi_n[i-1]) / dx
            else:
                dphi_dx = (phi_n[i+1] - phi_n[i]) / dx
            
            # Derivada segunda (centrado)
            d2phi_dx2 = (phi_n[i+1] - 2*phi_n[i] + phi_n[i-1]) / dx**2
            
            # Atualização explícita
            phi[n+1, i] = phi_n[i] + dt * (
                -U * dphi_dx + Gamma * d2phi_dx2 + source[i]
            )
        
        # Condições de contorno
        if boundary == 'periodic':
            phi[n+1, 0] = phi[n+1, -2]
            phi[n+1, -1] = phi[n+1, 1]
        elif boundary == 'dirichlet':
            phi[n+1, 0] = phi0(x[0]) if callable(phi0) else phi0
            phi[n+1, -1] = phi0(x[-1]) if callable(phi0) else phi0
        # Neumann pode ser adicionado conforme necessidade
        
    return x, t, phi

# Exemplo de uso: Comparação entre os três fenômenos
print("=== Comparação: Evolução Temporal dos Três Fenômenos ===\n")

# Parâmetros comuns
L = 100      # m
U = 0.1      # m/s
T_final = 500  # s
dx = 1.0     # m

# Condição inicial: pulso gaussiano
def phi0_gauss(x, x0=50, sigma=5):
    return np.exp(-(x - x0)**2 / (2 * sigma**2))

# Casos: (nome, Gamma, cor)
casos = [
    ('Momentum (ν)', 1.0e-6, 'blue'),    # Água: ν ≈ 10⁻⁶ m²/s
    ('Calor (α)', 1.4e-7, 'red'),        # Água: α ≈ 1.4×10⁻⁷ m²/s  
    ('Massa (D)', 2.0e-9, 'green')       # O₂ em água: D ≈ 2×10⁻⁹ m²/s
]

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for ax, (nome, Gamma, cor) in zip(axes, casos):
    # Resolver
    x, t, phi = solver_transporte_1d(
        phi0=phi0_gauss, U=U, Gamma=Gamma, S=0,
        L=L, T_final=T_final, dx=dx
    )
    
    # Plotar evolução em 4 instantes
    t_plot = [0, T_final/3, 2*T_final/3, T_final]
    for ti in t_plot:
        idx = np.argmin(np.abs(t - ti))
        ax.plot(x, phi[idx, :], label=f't = {t[idx]:.0f}s', color=cor, alpha=0.7)
    
    ax.set_xlabel('Posição x [m]')
    ax.set_ylabel(f'{nome} normalizado')
    ax.set_title(f'{nome}\nΓ = {Gamma:.1e} m²/s')
    ax.legend(fontsize=8)
    ax.grid(True, alpha=0.3)
    
    # Calcular Pe local
    Pe_local = U * L / Gamma
    ax.text(0.02, 0.98, f'Pe = {Pe_local:.1e}', 
           transform=ax.transAxes, fontsize=9,
           bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.3),
           verticalalignment='top')

plt.tight_layout()
plt.show()

display(Markdown(f"""
> **Análise dos resultados (🔬 Pós-Graduação):**
> 
> | Fenômeno | Γ [m²/s] | Pe = UL/Γ | Comportamento Observado |
> |----------|----------|-----------|------------------------|
> | Momentum | 1.0×10⁻⁶ | ~1.0×10⁷ | Pulso praticamente não se alarga (advecção pura) |
> | Calor    | 1.4×10⁻⁷ | ~7.1×10⁷ | Mesmo comportamento: advecção dominante |
> | Massa    | 2.0×10⁻⁹ | ~5.0×10⁹ | Difusão ainda mais desprezível |
> 
> ✅ **Conclusão:** Para escalas de rio (L ~ 100 m), a difusão molecular é 
> irrelevante para todos os três fenômenos. A dispersão hidrodinâmica (E ~ 1-100 m²/s) 
> é o mecanismo difusivo relevante em aplicações ambientais.
"""))
```
